## Problema 10: Pendulo simple

_No sería esta una materia de física si no usáramos como ejemplo al oscilador armónico en algún momento, así que saquemos eso del medio._

Como recordarán de física elemental, la ecuación que rige la dinámica de una masa suspendida sujeta a un punto fijo mediante una cuerda inextensible de longitud $\ell$ (i.e. la ecuación que del péndulo simple) es
\begin{equation}
    \frac{\mathrm{d}^2 \theta}{\mathrm{d}t^2} + \frac{g}{\ell} \mathrm{sen}(\theta) = 0,
\end{equation}
que para pequeños apartamientos del equilibrio ($\theta = 0$) puede aproximarse como
\begin{equation}
    \frac{\mathrm{d}^2 \theta}{\mathrm{d}t^2} + \omega^2 \theta = 0,
\end{equation}
donde llamamos $\omega^2 = g/\ell$.

Aunque la forma usual de plantear este problema es como uno de valores iniciales (a partir de la posición y velocidad iniciales), si suponemos que parte del reposo y que tiene un período $T$, puede plantearse también como un problema de valores de contorno
\begin{equation}
    \theta(0) = \theta_0 \qquad \wedge \qquad \theta(T) = \theta_0.
\end{equation}

También en este régimen de pequeñas oscilaciones, el período resulta $T=2\pi/\omega$ y conocemos la solución analítica, dada por
\begin{equation}
    \theta(t) = \theta_0 \cos(\omega t).
\end{equation}

Aprovecharemos este hecho para resolver nuestra primer ecuación diferencial numéricamente y poder estimar el error cometido$^\dagger$. Para ello, considerando pequeñas oscilaciones:

```python
def D2_primer_orden(N, h):
    D2 = np.zeros ( (N,N) )

    # Primera fila
    D2[0,0] = -1
    D2[0,1] = -1
    D2[0,2] =  1

    # Primera mitad de la matriz (DDFF adelantadas)
    for i in range(1, N//2):
        D2[i,i-1] =  1
        D2[i,i  ] = -1
        D2[i,i+1] = -1
        D2[i,i+2] =  1
    
    # Segunda mitad de la matriz (DDFF atrasadas)
    for i in range(N//2, N-1):
        D2[i,i-2] =  1
        D2[i,i-1] = -1
        D2[i,i  ] = -1
        D2[i,i+1] =  1
    
    # Última fila
    D2[N-1,N-3] =  1
    D2[N-1,N-2] = -1
    D2[N-1,N-1] = -1

    return 1/(2*h**2)*D2
```

# Diferencias finitas

## Aproximación asimétrica de primer orden para la primera derivada

Supongamos que tenemos una función suficientemente suave $f(x)$ cuyos valores conocemos sobre una cierta grilla equiespaciada con paso $\Delta x$, es decir, $x_k = x_0 + k \Delta x, \ \ 0 \le k \le N-1$, con $N$ la cantidad de puntos de la grilla. Llamemos $f_k$ a $f(x_k$).

Bajo estas condiciones, podemos estimar numéricamente su derivada primera utilizando, por ejemplo, el valor de $f$ en el punto de interés $x_j$ y un punto adicional. Naturalmente, la aproximación será, en general, más precisa si éste punto adicional se encuentra cerca de $x_j$.

  Para encontrar una posible aproximación a la derivada, usando dos puntos, intentemos utilizar información conocida en $x_j$ y $x_{j+1}$, es decir el punto de interés y el siguiente al mismo. Luego, utilizando el [teorema de Taylor](https://es.wikipedia.org/wiki/Teorema_de_Taylor) para escribir $f_{j+1}$, tenemos
\begin{equation}
    f_{j+1} = f_j + \frac{1}{1!} f'_j h + \frac{1}{2!} f''(\xi) h^2 \qquad x_j \le \xi \le x_{j+1}.  \tag{1}
\end{equation}
donde $h = \Delta x = x_{j+1} - x_{j}$. Despejando la derivada de la primer ecuación obtenemos la _diferencia finita adelantada_ (ya que usa información adelante del punto de interés) _de primer orden_,
\begin{equation}
    f'_j = \frac{f_{j+1}-f_j}{h} - f''(\xi_\mathrm{f}) h = \frac{f_{j+1}-f_j}{h} + \mathcal{O} (h) \qquad x_j \le \xi_\mathrm{f} \le x_{j+1} \tag{2}.
\end{equation}
Utilizamos $\mathcal{O}(h)$ para denotar un término orden $h$. Finalmente, el aproximante buscado resulta
\begin{equation}
    f'_j = \frac{f_{j+1}-f_j}{h},
\end{equation}
y conlleva, en general, un error de orden $h$ en la estimación de la derivada. De esta manera, en un entorno lo suficientemente chico alrededor de $x_j$, al disminuir $h$ en un factor 2 el error disminuirá también en un factor $2$. Por este motivo, se dice que el método es de primer orden. Noten, sin embargo, que el orden solo nos dice cómo se modifica (asintóticamente) el error a medida que la resolución de la grilla aumenta o disminuye, pero no nos permite calcular la magnitud del error (que dependerá, en este caso, de la magnitud de la derivada segunda en un entorno del punto).

De manera similar, utilzando expansiones de Taylor alrededor de $x_{j+2}$, $\dots$, $x_{j+p}$, y también $x_{j-1}$, $\dots$, $x_{j-q}$ pueden obtenerse aproximaciones de mayor orden. Veamos como hacer esto para la derivada segunda.

## Aproximación simétrica (centrada) de cuarto orden para la derivada segunda

Utilicemos ahora la misma idea para obtener un aproximador a la derivada segunda, utilizando información conocida en $x_{j-2}$, $x_{j-1}$, $x_j$, $x_{j+1}$ y $x_{j+2}$. Escribiendo expresiones de Taylor análogas a las de (1) tenemos

\begin{alignat}{10}
    f_{j-2} - f_j &= \ &-& \frac{ f^{(1)}_j}{1!} 2h &+& \frac{f^{(2)}_j}{2!}(2h)^2 &-& \frac{f^{(3)}_j }{3!}(2h)^3 &+& \frac{ f^{(4)}_j}{4!} (2h)^4 &-& \frac{f^{(5)}_j }{5!} (2h)^5 &+& \frac{1}{6!} f^{(6)}(\xi_{j-2}) (2h)^6 \qquad &x_{j-2} \le \xi_{j-2} \le x_{j-1}, \tag{3}\\
    f_{j-1} - f_j &= \ &-& \frac{f^{(1)}_j }{1!} h &+&  \frac{f^{(2)}_j}{2!} h^2 &-& \frac{f^{(3)}_j}{3!} h^3 &+& \frac{f^{(4)}_j}{4!} h^4 &-& \frac{f^{(5)}_j }{5!} h^5 &+& \frac{1}{6!} f^{(6)}(\xi_{j-1}) h^6 \qquad & x_{j-1} \le \xi_{j-1} \le x_j, \tag{4}\\
    f_{j+1} - f_j &= \ &&\frac{f^{(1)}_j }{1!} h &+& \frac{f^{(2)}_j}{2!}h^2 &+& \frac{f^{(3)}_j }{3!} h^3 &+& \frac{f^{(4)}_j }{4!} h^4 &+& \frac{f^{(5)}_j}{5!} h^5 &+& \frac{1}{6!} f^{(6)}(\xi_{j+1}) h^6 \qquad &x_j \le \xi_{j+1} \le x_{j+1}, \tag{5}\\
    f_{j+2} - f_j &= \ &&\frac{f^{(1)}_j }{1!} 2h &+& \frac{f^{(2)}_j}{2!} (2h)^2 &+& \frac{f^{(3)}_j }{3!} (2h)^3 &+& \frac{f^{(4)}_j }{4!} (2h)^4 &+& \frac{f^{(5)}_j }{5!} (2h)^5 &+& \frac{1}{6!} f^{(6)}(\xi_{j+2}) (2h)^6 \qquad &x_{j+1} \le \xi_{j+2} \le x_{j+2}, \tag{6}
\end{alignat}
donde utilizamos ahora la notación $f^{(i)}$ para denotar la $i$-ésima derivada.

Proponiendo ahora una combinación lineal de la forma $A(3) + B(4) + C(5) + D(6)$, y recordando que queremos obtener la mejor aproximación posible a la derivada segunda, esta combinación lineal debe verificar

\begin{align}
    \left(-2A - B + C + 2D \right) f^{(1)}_j h &= 0, \\
    \left( 2A + \frac{B}{2} + \frac{C}{2} + 2D \right) f^{(2)}_j h^2 &= f^{(2)}_j h^2, \\
    \left(-\frac{8}{6}A - \frac{1}{6} B + \frac{1}{6} C + \frac{8}{6} D \right) f^{(3)}_j h^3     &= 0, \\
    \left( \frac{16}{24} A + \frac{1}{24} B + \frac{1}{24} C + \frac{16}{24} D \right) f^{(4)}_j h^4  &= 0, \\
\end{align}
que resolviendo para $A$, $B$, $C$ y $D$, permite obtener
\begin{equation}
    A = \frac{-1}{12} \qquad B = \frac{4}{3} \qquad C = \frac{4}{3} \qquad D = \frac{-1}{12}.
\end{equation}
(pueden verificar este resultado [acá](https://www.wolframalpha.com/input/?i=%7B%7B-2%2C+-1%2C+1%2C+2%7D%2C+%7B2%2C+1%2F2%2C+1%2F2%2C+2%7D%2C+%7B-8%2F6%2C+-1%2F6%2C+1%2F6%2C+8%2F6%7D%2C+%7B16%2F12%2C+1%2F12%2C+1%2F12%2C+16%2F12%7D%7D+*+%7BA%2C+B%2C+C%2C+D%7D+%3D+%7B0%2C+1%2C+0%2C+0%7D)). Notemos que esta solución anula, además, los coeficientes correspondientes a la derivada quinta, como podemos ver reemplazando
\begin{equation}
    \left[-  \frac{32}{120} \left(\frac{-1}{12}\right) - \frac{1}{120} \left(\frac{4}{3}\right) + \frac{1}{120} \left(\frac{4}{3}\right) + \frac{32}{120} \left(\frac{-1}{12} \right) \right] f^{(5)}_j h^5 = 0.
\end{equation}
Como vieron en teóricas, este tipo de cancelaciones suceden frecuentemente para esquemas centrados.

Tenemos entonces que, por construcción,
\begin{equation}
   \left(f_{j-2} - f_j\right) \left[ \frac{-1}{12} \right] + \left(f_{j-1} - f_j\right) \left[ \frac{4}{3} \right] + \left(f_{j+1} - f_j\right) \left[\frac{4}{3} \right] +  \left(f_{j+2} - f_j\right) \left[ \frac{-1}{6}\right] = f^{(2)}_jh^2 + \mathcal{O}(h^6),
\end{equation}
y despejando obtenemos
\begin{equation}
    f''_j = \frac{-f_{j+2} + 16 f_{j+1} -30 f_j + 16 f_{j-1} - f_{j-2}}{12h^2} + \mathcal{O}(h^4).
\end{equation}

Vemos que, efectivamente, utilizando los valores de $f$ sobre 5 puntos, pudimos encontrar un aproximante centrado a su derivada segunda que resulta de 4to orden.

## Uso de aproximantes para construir aproximantes a derivadas superiores


Otra manera de obtener esquemas en diferencias finitas para derivadas superiores es usar recursivamente aproximantes conocidos. Por ejemplo, si conocemos la derivada primera de manera exacta en cada punto, podemos escribir una aproximación de primer orden a la derivada segunda de la siguiente manera
\begin{equation}
    f''_j = \frac{f'_{j+1} - f'_{j}}{h} + \mathcal{O}(h).
\end{equation}
Esto es, ni más ni menos, que la fórmula que vimos en $(2)$, aplicada sobre la derivada segunda. Podemos ver fácilmente que no es necesario conocer $f'$ de manera exacta, ya que hacerlo con orden $h^2$ alcanzará para que el error en toda la expresión anterior se mantenga de orden $h$.

Como demostrarán en el problema 1, el siguiente es un aproximante de segundo orden a la derivada primera
\begin{equation}
    f'_j = \frac{f_{j+1} - f_{j-1}}{2h} + \mathcal{O}(h^2),
\end{equation}
resulta en
\begin{equation}
    f''_j = \dfrac{ \left[ \frac{f_{j+2} - f_{j}}{2h} + \mathcal{O}(h^2) \right] - \left[ \frac{f_{j+1} - f_{j-1}}{2h} + \mathcal{O}(h^2) \right] }{h} + \mathcal{O}(h) = \frac{f_{j+2} - f_{j+1} - f_{j} + f_{j-1}}{2h^2} + \mathcal{O}(h).
\end{equation}
Si bien este no es un aproximante óptimo considerando que utiliza valores de la función sobre 4 puntos, ilustra la idea de cómo construir esquemas de diferencias finitas para derivadas superiores a partir de aquellos conocidos para la primer derivada, siendo una alternativa a la construcción mediante expansiones de Taylor o, como vieron en teóricas, el polinomio interpolador de Lagrange.

## Resolución matricial de problemas de contorno

En caso que nos interese resolver numéricamente un problema de valores de contorno lineal de primer orden, es decir,
\begin{align}
    \frac{\mathrm{d} u}{\mathrm{d} x} &= f(x)u + g(x), \qquad \qquad \mathrm{para} \quad 0 < x < L \\
    u(0) &= u_0,
\end{align}
con $f$ y $g$ conocidas, una forma de hacerlo es integrando (analítica o numéricamente) el miembro derecho. Otra forma de realizar esta tarea, más relacionada con los contenidos del curso, es transformar esta ecuación diferencial en una ecuación de diferencias, usando los aproximantes que vimos en los problemas previos. Por ejemplo, si discretizamos nuestro dominio usando $N$ puntos, $x_i = i \frac{L}{N-1}, \quad i=0, \dots, N-1$, y dado que conocemos $u_0$, podemos escribir una ecuación para $u_1$ usando diferencias finitas adelantadas como
\begin{equation}
    \frac{u_1 - u_0}{h} = f_1 u_1 + g_1,
\end{equation}
con $h=L/(N-1)$. Podemos despejar $u_1$ como
\begin{equation}
    u_1 = - \frac{g_1 + u_0/h}{f_1 - 1/h}.
\end{equation}
Siguiendo este procedimiento, podríamos hallar $u_2$ a partir de $u_1$ y así hasta $u_{N-1}$, aunque arrastrando un cierto error a cada paso.

Alternativamente, la misma idea puede utilizarse si reescribimos nuestro problema de contorno sobre una grilla discreta de la siguiente manera
\begin{equation}
    D\mathbf{u} + \mathbf{b} = F\mathbf{u} + \mathbf{g},
\end{equation}
con
\begin{equation}
    D = \frac{1}{h} \begin{pmatrix}
          1 &  0 & 0 & \dots &  0 & 0 \\
         -1 &  1 & 0 & \dots &  0 & 0 \\
         0 & -1 & 1 & \dots &  0 & 0 \\
         \vdots & \vdots & \vdots & \ddots & \vdots & \vdots \\
         0 &  0 & 0 & \dots & -1 & 1
    \end{pmatrix}, \qquad
    F = \begin{pmatrix}
    f_1    &      0 &  \dots &      0 \\
    0      &    f_2 &  \dots &      0 \\
    \vdots & \vdots & \ddots & \vdots \\
    0      &   0    &  \dots & f_{N-1}
    \end{pmatrix}, \qquad
    \mathbf{u} = \begin{pmatrix}
        u_1 \\ u_2 \\ \vdots \\ u_{N-2} \\ u_{N-1}
    \end{pmatrix}, \qquad
    \mathbf{b} = \frac{1}{h}\begin{pmatrix}
        - u_0 \\ 0 \\ 0 \\ \vdots \\ 0
    \end{pmatrix}, \qquad
    \mathbf{g} = \begin{pmatrix}
    g_1 \\g_2 \\ \vdots \\  g_{N-2} \\ g_{N-1} \\
    \end{pmatrix}.
\end{equation}

En particular, la matriz $D$ es la representación matricial del operador de diferenciación y $\mathbf{b}$ contiene las condiciones de contorno (escaladas) del problema. Luego, $\mathbf{u}$ puede hallarse como la solución al sistema lineal
\begin{equation}
    A \mathbf{u} = \mathbf{c},
\end{equation}
con $A = D-F$ y $\mathbf{c} = \mathbf{g} - \mathbf{b}$. Para esto pueden utilizarse métodos clásicos de álgebra lineal (reducción gaussiana, metodos iterativos, matriz inversa, etc.). 

Si bien cuando la cantidad de puntos de grilla es reducida esto funcionará aceptablemente, dado que las matrices involucradas poseen una gran cantidad de ceros (i.e. son [ralas](https://es.wikipedia.org/wiki/Matriz_dispersa) — _sparse_ en inglés—), trabajar con las matrices completas se volverá muy ineficiente a resoluciones altas. Para suplir esta falencia, Numpy (y muchas otras bibliotecas de técnicas numéricas) implementan subrutinas especiales para matrices ralas. En prácticas posteriores veremos como emplearlas eficientemente.

## Análisis de consistencia

Para estudiar la consistencia entre una ecuación diferencial y la correspondiente ecuación de diferencias queremos ver si, cuando disminuye el espaciamiento entre puntos, la ecuación de diferencias tiende a la ecuación diferencial. Naturalmente, la consistencia depende tanto de la ecuación diferencial bajo estudio como del esquema de diferencias propuesto.

Supongamos que nos interesa resolver la ecuación de advección 1D
\begin{equation}
    \frac{\partial f(x,t)}{\partial t} + c \frac{\partial f(x,t)}{\partial x} = 0,  \qquad 0 < t < T \quad \wedge \quad 0 < x < L.
\end{equation}
Proponemos sendas grillas de $N$ y $J$ puntos $t^n = n T/(N-1) = n \Delta t$ y $x_j = j L/(J-1) = j\Delta x$. Noten que ahora el subíndice denota puntos del espacio y el superíndice puntos en tiempo (no potenciación). Proponemos entonces la aproximar nuestra ecuación con el esquema
\begin{equation}
    \frac{f_j^{n+1}-f_j^{n-1}}{2\Delta t} + c \frac{f^n_{j+1} - f^n_{j-1}}{2\Delta x} = 0,
\end{equation}
donde, naturalmente, $f^{n}_{j} = f(x_j, t^n)$.

Sea ahora $F$ una solución analítica a la ecuación de advección. Reemplazando en el esquema de diferencias propuesto, obtenemos
\begin{equation}
    \frac{F(x_j, t^{n+1}) -F(x_j, t^{n-1})}{2\Delta t} + c \frac{F(x_{j+1}, t^{n+1}) - F(x_{j-1}, t^n)}{2\Delta x} = \delta F,
\end{equation}
donde $\delta F$ es el _**error de truncamiento**_ (a veces llamado _**error de consistencia**_) y es una medida de qué tan bien o mal la solución real satisface nuestra ecuación en diferencias. Si la aproximación es buena, la solución analítica $F$ debería satisfacerla exactamente para $\Delta t \to 0 \wedge \Delta x \to 0$, es decir, $\delta F \to 0$.

Podemos obtener un orden para $\delta F$ en función de la resolución de la discretización empleando desarrollos de Taylor en $t^n$ y $x_j$, es decir
\begin{equation}
\delta F = \frac{1}{2\Delta t} \left( \left[F + \frac{\partial F}{\partial t} \Delta t + \frac{1}{2}\frac{\partial^2 F}{\partial t^2} (\Delta t)^2 + \mathcal{O}(\Delta t)^3 \right]_{x_j, t^n} - \left[F - \frac{\partial F}{\partial t} \Delta t + \frac{1}{2} \frac{\partial^2 F}{\partial t^2} (\Delta t)^2 + \mathcal{O}(\Delta t)^3 \right]_{x_j, t^n} \right) + \frac{c}{2\Delta x} \left( \left[F + \frac{\partial F}{\partial x} \Delta x + \frac{1}{2}\frac{\partial^2 F}{\partial x^2} (\Delta x)^2 + \mathcal{O}(\Delta x)^3\right]_{x_j, t^n} - \left[F - \frac{\partial F}{\partial x} \Delta x + \frac{1}{2} \frac{\partial^2 F}{\partial x^2} (\Delta x)^2 + \mathcal{O}(\Delta x)^3 \right]_{x_j, t^n} \right).
\end{equation}

Reagrupando tenemos
\begin{equation}
\delta F = \left[ \frac{\partial F}{\partial t} + c \frac{\partial F}{\partial x} \right]_{x_j, t^n} + \mathcal{O}\left[(\Delta t)^2\right] + \mathcal{O}\left[(\Delta x)^2\right].
\end{equation}
El primer término del miembro derecho se anula ya que $F$ es solución a la ecuación de advección en todo punto, en particular, en $(x_j, t^n)$. Obtuvimos entonces
\begin{equation}
\delta F = \mathcal{O}\left[(\Delta t)^2\right] + \mathcal{O}\left[(\Delta x)^2\right],
\end{equation}
que tiende a $0$ para $\Delta t \to 0$ y $\Delta x \to 0$ y, entonces, el esquema sugerido es consistente. 

Noten que $\delta F \to 0$ no alcanza para garantizar que $f$ converja a $F$ sobre la grilla discreta. Además de consistente es necesario que el esquema sea estable para garantizar convergencia. Veremos más sobre esto en clases posteriores.
