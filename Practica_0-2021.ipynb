## **1.1 Tipos de datos básicos**
([Regresar a la tabla de contenidos](#contenidos))


Por defecto Python puede trabajar, entre otros que no serán de interés, con los siguientes tipos de datos básicos:
* **Números enteros**: los que ya conocen desde la primaria, el conjunto $(\dots, -3, -2, -1, 0, 1, 2, 3, \dots)$. Ejemplo: `2`.
* **Números de punto flotante**: son un subconjunto de los números racionales, aunque a la hora de trabajar es útil pensar que representan números reales. La manera en que se construye este conjunto escapa los objetivos de esta notebook, pero quienes estén interesades pueden ver una introducción en [Wikipedia](https://es.wikipedia.org/wiki/Coma_flotante). Si el número real con el que queremos trabajar no puede ser representado como un número de punto flotante (la gran mayoría), Python lo aproximará automáticamente al número de punto flotante más cercano. Ejemplo: `3.141592653589793115997963468544185161590576171875` (notar que esto es una aproximación de punto flotante a $\pi$). 

    También pueden expresarse utilizando notación científica de la manera `xey`, donde `x` es la mantisa e `y` es el exponente. Por ejemplo, la aproximación de punto flotante a $1.3\times10^{-5}$ podemos escribirla como `1.3e-5`.

* **Números complejos** (de punto flotante): Los números complejos en Python son un par de números de punto flotante, el primero denotando la parte real y el segundo la imaginaria. En los números complejos de Python, la unidad imaginaria se denota con j. Ejemplo: `5.6 + 3.2e-2j` es la aproximación de punto flotante al número $5.6 + 3.2\times10^{-2} i$. Noten que la expresión `j` no tiene sentido sintáctico, siempre debe acompañarse de un número. De esa manera, la unidad imaginaria se escribe `1j`.

* **Datos lógicos (o booleanos)**: útiles para organizar tareas complejas, siguen las reglas del álgebra booleano. El conjunto de datos lógicos tiene dos elemementos, `True` (verdadero) y `False` (falso).
* **Cadenas de caracteres (_strings_):** consisten en una sucesión de caractéres, como por ejemplo `abc` o `3r 0.78g\`. A la hora de definir strings debemos encerrar el contenido entre comillas simples o dobles, es decir, `"abc"` o `'3r 0.78g\'`.

Corran (¡`Ctrl+Enter`!) la siguiente celda para ver como Python entiende este tipo de datos (dado que es la primera operación que ejecutan deben aguardar unos segundos a que Google inicialice el kernel — pueden observar el estado del kernel arriba a la derecha—).

````python
print(3)             # Imprimo el entero 3
print(type(3))       # Imprimo el tipo de 3
print()              # Línea vacía

print(1.4142)        # Imprimo una aproximación a √2
print(type(1.4142))  # Imprimo una aproximación a √2
print()              # Línea vacía 

print("hola")        # Imprimo el texto hola
print(type("hola"))  # Imprimo el texto hola
print()              # Línea vacía
````

Como resultado, deberían ver impresos los números `3`, `1.4142` y el texto `hola`. Debajo de cada elemento imprimimos además el tipo de dato en cuestión, que lo podemos obtener usando la función `type`, obteniendo `<class 'int'>`, `<class 'float'>` y `<class 'str'>`, es decir un entero, un número de punto flotante y un string, respectivamente. La función `type` de vuelve el tipo de dato del argumento pasado entre paréntesis.

Noten que para imprimir algo en la pantalla usamos la función `print`, que imprime lo que nosotros le pasamos como argumento entre paréntesis, y en particular `print()` imprime una línea en blanco. También pueden pasarse varios argumentos a `print` en cuyo caso los concatena de izquierda a derecha e imprime el resultado.

Finalmente reparen en el texto escrito en verde. Noten que está precedido por un `#`. En Python, cualquier cosa escrita después de un signo `#` no es procesada durante la ejecución†. Este tipo de texto se denomina **comentario** y sirve para dar información sobre el código al lector del mismo. De manera relacionada, noten que en general los espacios no tienen sentido sintáctico. Una excepción a este último punto son los espacios en los strings (ej. `metodos numéricos`) o los **espacios en el comienzo de una línea** (cuyo significado veremos más adelante). 

†: Excepto si el signo `#` forma parte de un string, por ejemplo, `"ab#c"`.

### **1.1.1 Conversión de tipos**
En algunas ocasiones, pueden precisar convertir un dato de un tipo a otro. Por ejemplo, convertir el número entero `6` al número de punto flotante `6.0`, o el string `"14.356"` al número de punto flotante `14.356`. Esto puede lograrse sencillamente utilizando las siguientes funciones:

* `str(arg)`: convierte lo que le pasamos entre paréntesis (`arg`) a un string.
* `int(arg)`: convierte lo que le pasamos entre paréntesis (`arg`) a un número entero.
* `float(arg)`: convierte lo que le pasamos entre paréntesis (`arg`) a un número de punto flotante.
* `complex(arg)`: convierte lo que le pasamos entre paréntesis (`arg`) a un número complejo.
* `bool(arg)`: convierte lo que le pasamos entre paréntesis (`arg`) a un valor booleano.

El modo en que operan estas conversiones es bastante sencillo, salvo por `bool`, que devuelve `False` solo el argumento es 0 (para números) o eel string nulo `""`. Además, `int`, que naturalmente no funciona si el argumento es complejo, devuelve el entero más cercano a cero en caso que el argumento sea un número de punto flotante. Veremos más adelante para que nos puede servir la conversión de tipos.

## **1.2 Operaciones aritméticas**
([Regresar a la tabla de contenidos](#contenidos))

Realizar operaciones aritméticas en Python es bastante sencillo. Por defecto, para operar entre números enteros y/o números de punto flotante el lenguaje incluye soporte para la suma (`+`), resta (`-`), multiplicación (`*`), división (`/`) y potenciación (`**`). Noten que dividir dos enteros puede generar un número de punto flotante (es decir, la división no es una operación cerrada en los enteros).

Corran la siguiente celda para ver estas operaciones en acción.

````python
print("Resultado de 2 + 6:      ", 2    + 6   ) # Suma de enteros
print("Resultado de 2.0 + 6.0:  ", 2.0  + 6.0 ) # Suma de punto flotante 
print("Resultado de 2 + 6.0:    ", 2    + 6.0 ) # Suma mixta
print()

print("Resultado de 10 - 2:     ", 10   - 2  )  # Resta de enteros
print("Resultado de 10.0 - 2.0: ", 10.0 - 2.0)  # Resta de punto flotante 
print("Resultado de 10 - 2.0:   ", 10   - 2.0)  # Resta mixta
print()

print("Resultado de 4 * 2:      ", 4    * 2   ) # Multiplicación de enteros
print("Resultado de 4.0 * 2.0:  ", 4.0  * 2.0 ) # Multiplicación de pto flotante 
print("Resultado de 4 * 2.0:    ", 4    * 2.0 ) # Multiplicación mixta
print()

print("Resultado de 16 / 2:     ", 16   / 2  )  # División de enteros
print("Resultado de 16.0 / 2.0: ", 16.0 / 2.0)  # División de punto flotante 
print("Resultado de 16 / 2.0:   ", 16   / 2.0)  # División mixta
print()

print("Resultado de 2**3:       ", 2   ** 3   ) # Potenciación de enteros
print("Resultado de 2.0**3.0:   ", 2.0 ** 3.0 ) # Potenciación de punto flotante 
print("Resultado de 2**3.0:     ", 2   ** 3.0 ) # Potenciación mixta
````

Al correr la celda deberían ver el resultado de las operaciones. Noten que no hay problema en mezclar enteros y números de punto flotante en una misma operación. Python _promueve_ el entero a un número de punto flotante y luego opera.

Adicionalmente cuando se trabaja con enteros pueden ser útiles las operaciones de cociente y resto de una división. La primera se denota con `//` mientras que la segunda con `%`. Veamos un ejemplo.

````python
print("Cociente de dividir 4 por 3:", 4 // 3 )
print("Resto de dividir 4 por 3:   ", 4  % 3 )
print()

print("Cociente de dividir 4.0 por 3.0:", 4.0 // 3.0)
print("Restp de dividir 4.0 por 3.0:   ", 4.0  % 3.0)
````

Noten que, además que si bien resultan más lógicas de introducir con operandos enteros, pueden usarse igualmente con números de punto flotante.

Para realizar operaciones encadenadas, Python establece un orden de precedencia en sus operaciones. En este [enlace](https://entrenamiento-python-basico.readthedocs.io/es/latest/leccion3/operadores_aritmeticos.html) pueden encontrar el orden de precedencia que usa Python para las operaciones aritméticas. Como verán, Python sigue el orden de operaciones que aprendieron en la primaria. Una alternativa a dejar que Python determine el orden de las operaciones es, también al igual que aprendieron en la primeria, usar paréntesis. Veamos un ejemplo de como establecer el orden de operaciones de esta forma:

````python
print("El resultado de 2+(3*3) es:",   2+3*3)
print("El resultado de (2+3)*3 es:", (2+3)*3)
print("El resultado de 2+3*3 es:  ",   2+3*3)
````

Veamos finalmente que algunas de estas operaciones elementales pueden operar entre strings. Para ello, usemos el siguiente ejemplo

````python
print("hola"  + " mundo")
print("hola " * 3 )
````

Como vemos en el ejemplo, con `+` se puede operar entre 2 strings resultando en la concatenación de los mismos. Por otra parte, también puede _multiplicarse_ un string una cantidad entera de veces, generando repeticiones del mismo.

Si queremos construir un string que incluya un número entero, además de escribirlo entre `""`, podemos usar las funciones de conversión de tipos que vimos antes. De esta manera podemos concatenar números (que más adelante van a ser cantidades que no conocemos a priori) y strings. Veamos un ejemplo:

````python
camiseta = 5
frase = "E. Ginóbili vestía la camiseta " + str(camiseta) + " de la selección."
print(frase)
````

### Suma de booleanos
En Python, `True` y `False` se representan internamente como `1` y `0`, respectivamente, de forma que la lógica subyacente en un programa sea más legible. Es por ello, sin embargo, que las operaciones aritméticas no resulta cerradas, ya que operan sobre los enteros subyacentes, como muestra el siguiente ejemplo.

````python
print(True  + False)
print(True  +  True)
print(False -  True)
print(True  * False)
````

## **1.3 Operaciones lógicas y comparaciones**
([Regresar a la tabla de contenidos](#contenidos))

Otro conjunto de operaciones posibles que podemos realizar con los tipos de datos presentados son las **operaciones lógicas** y las **comparaciones**.

Las **operaciones lógicas** comprenden a la [conjunción lógica](https://es.wikipedia.org/wiki/Conjunci%C3%B3n_l%C3%B3gica#Definici%C3%B3n) (`and`), la [disyunción lógica](https://es.wikipedia.org/wiki/Disyunci%C3%B3n_l%C3%B3gica#Definici%C3%B3n) (`or`) y la [negación lógica](https://es.wikipedia.org/wiki/Negaci%C3%B3n_l%C3%B3gica#Definici%C3%B3n) (`not`) y operan sobre datos de tipo booleano, y el resultado de la operación es del mismo tipo. Las tablas de verdad de estas operaciones son las que aprendieron en álgebra, aunque las pueden consultar en los enlaces.

Un ejemplo de estas operaciones (que también pueden ordenarse usando paréntesis) es el siguiente

````python
print("Verdadero Y Falso:     ",       True and False )
print("Verdadero O Falso:     ",       True  or False )
print("(No Verdadero) Y Falso:", (not True) and False )
print("No (Verdadero Y Falso):",  not (True and False))
````

Los operadores de comparación incluyen a las desigualdades `<` (menor que), `>` (mayor que), `<=` (menor o igual que), `>=` (mayor o igual que), `==` (igual a) y `!=` (distinto a). Noten que, dado que `=` denotaba la operación de asignación, es necesario utilizar otro símbolo para el operador de _igual a_, por lo que se usa un doble signo igual `==`. Naturalmente, el resultado de una operación de comparación es un dato booleano (`True` o `False`).

El uso más común de estas relaciones es el de comparar números (sean estos enteros o de punto flotante), aunque pueden también aplicarse a otros tipos de datos. Analizaremos el resultado de utilizar estos operadores sobre datos no numéricos cuando resulte necesario.

### Sobre la igualdad entre números de punto flotante


Finalmente, como regla general, vale remarcar que **suele ser una muy mala práctica comparar igualdad entre números de punto flotante**. Sin embargo, al trabajar con aritmética de punto flotante, puede suceder que se pierda precisión en alguno de los bits menos significativos, y por lo tanto el operador `==` devolverá falso en casos donde esto significa un error.

Veamos un ejemplo

````python
print(0.1 + 0.2 == 0.15 + 0.15)
````

Como vemos, Python está reportando que 0.3 no es igual a 0.3. Suele ser más apropiado considerar a dos números de punto flotante como iguales si su distancia es menor a una cierta tolerancia. ### **1.6.1 Expresiones condicionales**

Las expresiones condicionales nos permiten escribir programas que ejecuten una u otra instrucción dependiendo del valor de verdad de una proposición (la condición). Esta es una de las principales utilidades de los datos booleanos y las operaciones de comparación que vimos anteriormente.

El símbolo utilizado para denotar el comienzo de un bloque de ejecución condicional es el de `if (condición):`. En caso que lo que se busque es ejecutar una instrucción si la condición es verdadera y otra si es falsa, contamos con la palabra clave `else:`, que antecede al bloque de instrucciones a ejecutar si la condición es falsa.

Veamos como funciona con un ejemplo

````python
def divisible_por_19(num):
    resto = num % 19  # Resto de dividir al número por 19.

    if resto == 0:    # Si el resto es cero, reportamos que es divisible.
        print(num, "es divisible por 19.")
    else:             # Caso contrario, reportamos que no es divisible.
        print(num, "no es divisible por 19.")

divisible_por_19(35)
divisible_por_19(20634)
````

Veamos que para denotar las instrucciones que deben ejecutarse condicionalmente, usamos un mayor nivel de indentación, al igual que sucedía con las funciones.

En caso que se busque tener más de dos opciones de ejecución posibles, Python cuenta con la instrucción `elif (condicion):`. Su uso es bastante intuitivo, como podemos ver en el siguiente ejemplo:

````python
# Función que reporta si un número a es mayor, menor o igual que otro b
def mayor_menor_igual(a, b):
    if a > b:         # Si es mayor
        print(a, "es mayor que", b)
    elif a < b:       # Si es menor
        print(a, "es menor que", b)
    else:             # Si no se cumplieron las condiciones anteriores
        print(a, "es igual a", b)

mayor_menor_igual(8  , 4*2)
mayor_menor_igual(9  , 3*7)
mayor_menor_igual(5*3, 2*2)
````

### **1.6.2 Bucles**

Hasta ahora no contamos con ningún mecanismo que nos permita realizar una tarea una cantidad arbitraria de veces. Una forma de lograr esto es utilizando bucles para evitar repetir una y otra vez la misma instrución (o una muy similar). En Python hay dos tipos de bucles, de tipo `while` o de tipo `for`. 

<a name="while"></a>
#### **1.6.2.1 Tipo `while`**

La sintaxis para los primeros es `while (condicion):`, y continuan ejecutando una instrucción o conjunto de instrucciones hasta que se verifique la condición requerida. Son útiles en cálculo numérico para, por ejemplo, ejecutar un algoritmo hasta alcanzar un cierto nivel de precisión (tolerancia). 

Por ejemplo, si queremos encontrar una raíz de la función $f(x) = a x^2 + bx + c$ en el intervalo $[x_0, x_1]$ para una dada tolerancia `tol` usando el [método de bisección](https://es.wikipedia.org/wiki/M%C3%A9todo_de_bisecci%C3%B3n) podemos usar el siguiente algoritmo

````python
def raiz_cuadrática(a, b, c, x0, x1, tol):
    f0 = a*x0**2 + b*x0 + c  # Función en el extremo izquierdo
    f1 = a*x1**2 + b*x1 + c  # Función en el extremo derecho
    
    #  Itero hasta alcanzar la tolerancia deseada
    #  Dado que todavía no vimos cómo calcular valores absolutos
    # vamos a comparar errores cuadráticos.
    while (f0**2 > tol**2) and (f1**2 > tol**2):
        # Busco el punto medio del intervalo y el valor de la función allí
        xi = (x0 + x1)/2
        fi = a*xi**2 + b*xi + c

        # Decido si xi debe reemplazar a x0 o a x1 como extremo de intervalo
        if f0*fi >= 0:
            x0 = xi
            f0 = fi          
        elif f1*fi >= 0:
            x1 = xi
            f1 = fi

    # ------------------------------------------------
    # Llegado este punto, el método terminó de iterar.
    
    #  Devuelvo el valor adecuado de acuerdo a en qué extermo se verifica
    # la condición de convergencia.
    if f0**2 < tol**2:
        return x0
    
    elif f1**2 < tol**2:
        return x1


raiz = raiz_cuadrática(138, 19682, 765, -200, -5, 1e-8)
print("Una raíz del polinomio 138x² + 19682x + 765 en el intervalo [-200,-5] ",
      "es:", raiz)
print("El valor del polinomio allí es", 138*raiz**2 + 19682*raiz + 765)
````

Hay varias cosas para analizar en el ejemplo anterior, que integra varias de los elementos sintácticos que hemos discutido. Lo primero a notar es el bloque `while` que se encarga de llevar a cabo la bisección del intervalo tantas veces como sea necesario hasta alcanzar una raíz, condición que especificamos con `(f0**2 > tol**2) and (f1**2 > tol**2)`. Notar que esta condición es equivalente a $|f_0|>\mathrm{tol} \ \mathrm{y} \ |f_1|>\mathrm{tol}$.

Noten además que, tanto para reconocer en qué extremo del intervalo se halla la raíz utilizamos bloques `if`, al igual que para analizar los casos límites antes de empezar a iterar. A su vez, los valores de retorno de la función también varían para distintos bloques `if`, por lo que varios de ellos tienen su propia instrucción `return`. Noten que cuando se ejecuta una instrucción `return`, Python inmediatamente abandona la ejecución, puesto que ya obtuvo el valor de retorno buscado.

#### **1.6.2.2 Tipo `for`**

Noten que en numerosas ocasiones puede ser deseable iterar una cantidad fija de veces en lugar de hacerlo hasta alcanzar una condición. Si bien resulta sencillo rescribir este escenario como una condición condición para un bucle `while` (¡lo hiceron en la última celda!), es también un escenario propicio para introducir los más populares bucles de tipo `for`.

En un bucle de tipo `for`, se recorre un iterable† y en cada iteración una variable va tomando el valor subsiguiente del iterable. Por ello la sintáctica de un bucle de este tipo es `for variable in iterable:`. Un iterable bastante sencillo de construir es aquel dado por `range(n)`, con `n` un número natural, que devuelve el conjunto de números `(0, 1, ..., n-1)`.

Veamos como podemos usar un bucle `for` para iterar un máximo de `N` veces el método de bisección anterior (donde ahora que ya están cancheres con el mismo, agregamos el chequeo de errores y de casos límites).

````python
def raiz_cuadrática(a, b, c, x0, x1, N):
    f0 = a*x0**2 + b*x0 + c  # Función en el extremo izquierdo
    f1 = a*x1**2 + b*x1 + c  # Función en el extremo derecho

    # Verifico que el método sea aplicable.
    if f0 * f1 > 0:
        print("No hay cambios de signo en el intervalo. No puede usarse el",
              "método de bisección.")
        return
    
    #  Considero los casos donde se anula en los extremos propuestos.
    #  Dado que todavía no vimos cómo calcular valores absolutos
    # vamos a comparar errores cuadráticos.
    elif f0**2 < tol**2:
        return x0
    elif f1**2 < tol**2:
        return x1
    
    #  Si no se cumplió ninguno de los casos anteriores, empiezo a iterar
    # el método.  
    else:
        for i in range(N):
            print("Realizando la iteración", i)

            # Busco el punto medio del intervalo y el valor de la función allí
            xi = (x0 + x1)/2
            fi = a*xi**2 + b*xi + c

            # Decido si xi debe reemplazar a x0 o a x1 como extremo de intervalo
            if f0*fi >= 0:
                x0 = xi
                f0 = fi          
            elif f1*fi >= 0:
                x1 = xi
               f1 = fi

    # ------------------------------------------------
    # Llegado este punto, el método terminó de iterar.
    
    #  Devuelvo el valor adecuado de acuerdo a en que extermo se verifica
    # la condición de convergencia.
    if f0**2 < f1**2:
        return x0  
    else:
        return x1



raiz = raiz_cuadrática(138, 19682, 765, -200, -5, 25)
print("Al cabo de 25 iteraciones, una raíz del polinomio 138x² + 19682x + 765 ",
      "en el intervalo [-200,-5] es:", raiz)
print("El valor del polinomio allí es", 138*raiz**2 + 19682*raiz + 765)
````

Como vemos en el ejemplo, la instrucción de bisección se realizó exactamente 25 veces, como indicamos al invocar la función.

Se diferencian de las listas en que una vez definidas sus elementos no pueden cambiar. Esto ayuda a evitar modificar accidentalmente (por ejemplo por un error de programación) cantidades que deben permanecer fijas. Las mismas se definen usando la función `tuple` o bien rodeando sus elementos por paréntesis, es decir, `tupla = (1, 5, "abc", 0.33)`. Luego, la instrucción `tupla[1] = 3.14` genera un error. Las tuplas tampoco permiten agregar nuevos elementos.


* **Diccionarios**: representan un conjunto de datos no ordenado secuencialmente. Pueden tenerse diccionarios de cualquiera de los tipos mencionados anteriormente (o de una combinación de ellos). Dado que los mismos no están ordenados, cada elemento del diccionario tiene un nombre o llave (`key` en inglés). Para su definición puede usarse la función `dict` o bien encerrar pares `nombre: valor` entre llaves (`{}`). Ejemplo: `newton = {"nombre": "Isaac", "apellido": "Newton", "fallecimiento": 84}`. Luego, de manera similar a como usábamos índices para acceder a los elementos de una lista, en un diccionario usamos los nombres de los distintos campos para acceder a los correspondientes valores. Si quisiéramos acceder a la edad de Sir Isaac al momento de fallecer, bastaría con la sintaxis `newton["fallecimiento"]`. Los elementos de un diccionario son mutables, es decir, pueden cambiarse dinámicamente. Por ejemplo, para agregar un nuevo valor, por ejemplo un título honorífico, basta con la expresión `newton["titulo"] = "Sir"`. Por otra parte, si ahora quisieramos expresar la fecha de su fallecimiento y no su edad en dicho momento, alcanza con la instrucción `newton["fallecimiento"] = "31 de marzo de 1727"` para lograrlo.

Vale aprovechar la ocasión para remarcar que no es casual que `range`, la función que usamos en nuestros primeros bucles `for`, comience en 0 y sea no inclusiva (`range(n)` termina en `n-1`). **Python**, al igual que muchos otros, **es un lenguaje basado-en-0**. De esta manera, para imprimir los 4 elementos de nuestra lista, podemos escribir
```python
for i in range(4):
    print(lista[i])
```

Veamos con un ejemplo como podemos crear una lista con los divisores primos de $N$.

````python
def es_primo(num):            # Función que determina si num es primo o no
    for i in range(2, num):   # Pruebo todos los posibles divisores hasta num
        if num%i == 0:        # Si encuentro un divisor, no es primo
            return False
    
    # Si el bucle se ejecutó hasta el final, num no tiene divisores
    return True

def divisores_primos(N):
    divisores = []           # Creo una lista vacía
    for i in range(2, N+1):  # Empiezo a probar divisores (+1 por si es primo)
        if es_primo(i):          # Solo con posibles divisores primos
            while N%i == 0:          # Mientras i siga siendo divisor
                divisores = divisores + [i]  # Guardo el divisor
                N = N / i                     # Alcanza ahora con mirar
                                              # divisores de N/i  
    return divisores

print("Divisores de 7:   ", divisores_primos(7)   )
print("Divisores de 56:  ", divisores_primos(56)  )
print("Divisores de 9875:", divisores_primos(9875))
````


Primeramente vale notar que con vistas a disminuir la complejidad del ejemplo, el algoritmo empleado es bastante ineficiente para valores grandes de $N$. Esto no es relevante en el contexto de este documento.

En cuanto a nuevas funciones (o usos diferentes de las mismas), noten que ahora le pasamos dos argumentos a la función `range`. Esta función, al igual que muchas otras, modifica levemente su comportamiento dependiendo de la cantidad de argumentos que recibe. Ya vimos que sucede cuando le pasamos un solo argumento. Cuando recibe dos, el primero es el primer elemento que usamos para iterar (`2` en `es_primo`), mientras que el segundo (`num` en `es_primo`) prescribe cuando detener la iteración. Recuerden que `range` es no inclusivo (`range(n)` termina en `n-1`). Otro elemento en el cual reparar es que `print` puede también imprimir listas.

También concatenamos listas usando el operador `+`. La línea `divisores = divisores + [i]`, crea una lista con `i` como único elemento y la concatena con la lista divisores existente, el resultado pasa a ser la nueva lista de divisores.

### 1.7.1 Múltiples valores de retorno y asignaciones simultáneas

Como vimos más arriba, las funciones pueden devolver listas, tuplas y diccionarios) sin problemas. Alternativamente, en caso que la lista o tupla a devolver tenga unos pocos valores, puede sencillamente separarse a los mismos con comas y Python construye automáticamente una tupla para retornar. Es decir
```python
def prueba():
    valor1 = 1
    valor2 = 0.5
    return valor1, valor2

tupla_resultado = prueba()
```

Si bien esto completamente equivalente a construir manualmente la tupla `tupla = (1, 0.5)` y utilizar `return tupla`, la estrategia de separar valores de retorno con comas es tan extensa que merece una mención.

De manera similar, también puede usarse la notación de separar variables con comas en las operaciones de asignación. De esta manera, suponiendo que la tupla `valores` contiene 3 valores, podemos obtener variables que almacenen a cada uno de ellos individualmente mediante
```python
valor1, valor2, valor3 = valores.
```

Veamos este [azúcar sintático](https://es.wikipedia.org/wiki/Az%C3%BAcar_sint%C3%A1ctico) en operación, calculando una aproximación a Pi y el respectivo error mediante la [fórmula de Leibniz](https://es.wikipedia.org/wiki/Serie_de_Leibniz).

````python
def leibniz(N):
    pi = 0                         # Inicializo mi aproximación en 0
    for n in range(0, N):          # Empiezo a recorrer los términos de la serie
        pi = pi + 4* (-1)**n / (2*n +1)  # Actualizo mi estimación de pi
    
    error = (pi - 3.1415926535897931159)**2  # Calculo el error

    return pi, error   # Devuelvo una tupla  (pi, error)

print("La función leibniz devuelve algo de tipo:", type(leibniz(5)))
print()

aproximacion, error_cuadratico = leibniz(20000)

print("Usando 20000 términos, la serie de Leibniz produce:", aproximacion)
print("El error cuadrático es de:", error_cuadratico)
````

Como vemos, la función devuelve una tupla.

Además, resulta sencillo **_desempaquetar_** cada valor de la tupla en una variable, lo que hicimos con
```python
aproximacion, error_cuadratico = leibniz(20000)
```

Además aprendimos que la tasa de convergencia de la fórmula de Leibniz es paupérrima.

<a name="slices"></a>
### 1.7.2 Rebanadas (slices) e índices negativos


En el caso de las listas y tuplas hay un mecanismo sencillo para acceder a un subconjunto de elementos de manera "simultánea". Para ello es necesario que los índices que se buscan acceder sigan un patrón lineal, es decir, sea del estilo
```
    Empezando en el elemento I quiero seleccionar hasta el elemento F-1 usando un paso P.
```
El mecanismo que nos permite realizar esta tarea es el de tomar rebanadas (o slices) de la lista. Para ello, basta con pasar como argumentos `I`, `F` y `P` pero separados por `:` en lugar de por comas. En caso de omitir `P` se considera que el paso es 1, mientras que en caso de omitir `I` se considera que la selección empieza en el primer elemento. Análogamente si se omite `F` la selección abarca hasta el último elemento.

Veamos algunos ejempos

````python
lista = list(range(10))

print("La lista es:                       ", lista       )
print("La rebanada '5:' de la lista es:   ", lista[5:]   )
print("La rebanada ':3' de la lista es:   ", lista[:3]   )
print("La rebanada '3::2' de la lista es: ", lista[3::2] )
print("La rebanada '3:8:2' de la lista es:", lista[3:8:2])
print("La rebanada '::4' de la lista es:  ", lista[::4]  )
````

Noten que en caso que se omita, por ejemplo, el paso, puede obviarse el segundo `:`. No obstante, en caso que se omita el elemento inicial o el final no puede realizarse lo mismo con la primer aparición de `:`.

Además de las rebanadas, otra manera conveniente (y no excluyente) de referirse a los elementos de una lista es utilizando índices negativos. Los índices negativos comienzan a contar desde el final, de esta manera, el índice `-1` denota el último elemento, mientras que `-3` hace lo propio para el antepenúltimo elemento.

Adicionalmente, los índices negativos también pueden combinarse con la notación de rebanadas, como podemos ver en el siguiente ejemplo. 

````python
lista = list(range(10))

print("La lista es:                        ", lista       )
print("El elemento -7 de la lista es:      ", lista[-7]   )
print("La rebanada '-8:' de la lista es:   ", lista[-8:]  )
print("La rebanada '-6:-1' de la lista es: ", lista[-6:-1])
print("La rebanada '-4::2' de la lista es: ", lista[-4::2])
print("La rebanada '::-4' de la lista es:  ", lista[::-4] )
````

Como podemos ver, mezclar índices negativos con con rebanadas resulta bastante sencillo. En particular, vale resaltar que el uso de un paso negativo genera un subconjunto cuyo orden está invertido con respecto a la aparición de los elementos en la lista original.

Finalmente, es pertinente aclarar que aún cuando todos los ejemplos mostrados utilizaron listas, tanto las rebanadas como los índices negativos pueden ser igualmente aplicados a tuplas.

### 1.7.3 Cantidad de elementos en listas y diccionarios

No sabemos como obtener la cantidad de elementos de una lista/tupla o diccionario. En el caso de las listas, una manera de conocer cuantos elementos tiene es mediante la función `len`, que devuelve justamente la cantidad de elementos de una lista. En cuanto a los diccionarios, una manera posible de hacer lo propio es mediante el siguiente procedimiento de dos etapas: primero obtenemos una lista con todos los elementos (perdiendo de esta manera las referencias a sus nombres/llaves), mediante la función `values`, y luego, sobre esa lista, podemos usar el ya conocido `len`, como demuestra el siguiente ejemplo.

````python
diccionario = {"Pi": 3.1415, "sqrt(2):": 1.4142, "e": 2.7182}

elementos = list(diccionario.values())
print("Los elementos del diccionario son: ", elementos)

for i in range(0, len(elementos)):
    print("El elemento ", i, "es: ", elementos[i])
````

## **2.2 Creación de arreglos**
([Regresar a la tabla de contenidos](#contenidos))

Existen numerosas maneras de crear arreglos en `numpy`. A continuación detallamos algunas de las que usaremos frecuentemente en el curso:

* **A partir de una lista**: una opción para crear un arreglo es utilizando una lista que contenga los datos buscados y la función `array` de `numpy`. Alternativamente, en caso de buscar crear una matriz se pueden usar listas de listas, de la siguiente manera `matriz = np.array([[1, 2], [3, 4]])`.

* **Cambiando la dimensión de un arreglo existente**: si ya contamos con un arreglo con la información buscada, pero el mismo tiene una dimensión diferente a la que deseamos, podemos redimensionalizar el mismo utilizando la función `reshape`. Por ejemplo, para convertir el vector `v` de 4 elementos en una matriz `m` de 2x2 podemos utilizar `m = np.reshape(v, (2, 2))`. Noten que para especificar la nueva forma del arreglo debemos utilizar una tupla con tantos elementos como dimensiones tenga el nuevo arreglo.

* **Números enteros comenzando en $I$, terminando en $F-1$ con paso $P$**: para crear un vector de números enteros comenzando en $I$ , terminando en $F-1$ y con paso $P$ podemos usar la función `arange` (noten la similitud con `range`), de la siguiente manera: `secuencia = np.arange(I, F, P)`.

* **$N$ puntos equiespaciados entre $x_0$ y $x_1$**: en caso que busquemos generar una **grilla** equiespaciada entre $x_0$ y $x_1$ utilizando $N$ puntos podemos usar la función `linspace` (i.e. linealmente espaciados). Una grilla de 256 puntos entre 0 y 1 puede construirse mediante `x = np.linspace(0, 1, 256)`.

* **Arreglo de ceros**: para construir un arreglo nulo con dimensiones (2, 2) (es decir, una matriz de 2x2) podemos utilizar la función `zeros`. De esta manera, `nula = np.zeros((2,2))`.

* **Arreglo de unos**: de manera similar, para construir un arreglo de unos con dimensiones (2, 2) podemos utilizar la función `ones`. De esta manera, `unos = np.ones((2,2))`.

* **Matriz identidad**: para construir un arreglo que represente una matriz identidad de $n\times n$ podemos usar la función `eye` de la siguiente manera `identidad = np.eye(n)`.

* **Arreglo de números al azar**: en caso de querer formar un arreglo con números al azar en el intervalo $[0,1)$ de dimensión arbitraria, `numpy` cuenta con la función `random.random_sample`†. Por ejemplo, para un arreglo de dimensión 3 de tamaño $m \times n \times q$ podemos expresar `azar = np.random.random_sample((m, n, q))`.

Como verán, en todos los casos antes del nombre de la función debemos colocar `np.`, es decir, el nombre interno que le pusimos a `numpy`. Esto es obligatorio en Python, ya que al invocar una función que no sea definida por nosotros mismos o incluida en Python por defecto, debemos anteceder la función que estamos invocando con el nombre de la correspondiente bibliteca, y separarlos con un `.`. Veamos algunos ejemplos de estas funciones.

†: En rigor, la función es `random_sample` y se halla dentro del submódulo `random` de Numpy, al que accedemos de la forma `np.random`. En caso que no resulte claro, no se preocupen por la distinción entre estos conceptos.

````python
import numpy as np  # Importamos numpy

identidad        = np.eye(3)                  # Creo una matriz identidad de 3x3
identidad_vector = np.reshape(identidad, (9)) # La convierto en un vector

print("identidad        = \n", identidad       )
print()
print("identidad_vector = \n", identidad_vector)

print("-----------------------------------")
azar  =  np.random.random_sample((4,4))  # Matriz de números aleatorios en [0,1)
unos  =  np.ones((2,2))                  # Matriz de unos

print("azar = \n", azar)
print()
print("unos = \n", unos)

print("-----------------------------------")
de_lista      = np.array([ 3.3, 4.5, 5.7 ])
de_range      = np.arange(2, 11, 2)
equiespaciado = np.linspace(1.5, 5.5, 5)

print("de_lista       = ", de_lista)
print("de_range       = ", de_range)
print("equiespaciados = ", equiespaciado)
````

## **2.3 Asignación de arreglos**
([Regresar a la tabla de contenidos](#contenidos))

Asignar un valor a un elemento dado de un arreglo es extremadamente sencillo. El principio es similar a la asignación de un elemento de una lista, excepto que para arreglos hay que especificar todos los índices necesarios. Por ejemplo, para guardar el valor `3` en la 2da. fila y 3ra. columna del arreglo bidimensional `A`, basta con la expresión `A[1, 2] = 3`.

Veamos algunos ejemplos de esto:

````python
import numpy as np
vector = np.zeros((3))

print("Vector inicializado:", vector)
print("Tamaño de vector:", vector.shape)
for i in range(vector.shape[0]):
    vector[i] = i*5
print("Vector asignado:", vector)

print("-------------------------------")

matriz = np.zeros((2,4))
print("Matriz inicializada:\n", matriz)
print("Tamaño de matriz:", matriz.shape)

for i in range(matriz.shape[0]):
    for j in range(matriz.shape[1]):
        matriz[i,j] = 3*i + j + 10
print("Matriz asignada:\n", matriz)
````

Como prometimos, resulta muy sencillo asignar elementos en un arreglo `numpy`. Noten además que introdujimos una nueva propiedad de los arreglos, la propiedad `shape` (forma) que nos devuelve la longitud del arreglo en cada dimensión como una tupla. Luego `arr.shape[n]`, nos da la longitud en la `n`-ésima dimensión del arreglo `arr`. Para un arreglo bidimensional (una matriz) con `n=0` tenemos la cantidad de filas y `n=1` la cantidad de columnas.

La sintaxis `variable.propiedad` para acceder a una propiedad de la variable es frecuente en Python. Los arreglos tienen otras propiedades que podemos acceder de esa manera, como `size` (la cantidad total de elementos del arreglo) o `dtype` (el tipo de elementos que conforman el arreglo, por ejemplo `int64` para enteros de 8 bytes). También pueden accederse de esta forma a algunos _métodos_ (recuerden la disgresión en la sección [1.7.3](#len)) que operan directamente sobre el arreglo. Veremos algunos ejemplos en la próxima sección.

<a name="numpy_slices"></a>
#### **2.4.1 Rebanadas**

Además de seleccionarse mediante un único índice, y al igual que las listas, los arreglos también pueden indexarse utilizando rebanadas. Por ejemplo, para una matríz `A`, la instrucción `A[::2, 4:8]` selecciona fila por medio y las columnas 5ta a 9na de `A` (recuerden que Python comienza a contar por el 0). También pueden usarse índices negativos, de manera que `A[-1, -2:]` devuelve las última fila y las últimas 2 columnas de `A`. Veamos que esto efectivamente es así.

````python
import numpy as np

A = np.reshape(np.arange(0, 16), (4,4))

print("A   =\n", A)
print()
print("A[2:,:]=\n", A[2:, :])
print()
print("A[1::2, -3:-1]=\n", A[1::2, -3:-1])
print()
print("A[2:, ::3]=\n", A[2:, ::3])
````

Noten que en caso que queramos retener todas las filas (o columnas) basta con colocar `:` en el lugar correspondiente al primer índice (o el segundo).

Adicionalmente, pueden utilizarse rebanadas en las asignaciones. Por ejemplo, si queremos guardar el valor `10` en la 3ra fila de una matriz `A`, basta con la instrucción `A[2, :] = 10`.

También es posible asignar a un arreglo utilizando otro arreglo. Para ello, los arreglos a uno y otro lado del igual deben poseer igual dimensión. Por ejemplo, si `A` es una matriz de 3x3 y quiero guardar en la última columna el vector `[0, 1, 2]`, alcanza con expresar `A[:, -1] = np.arange(3)`.

Pongámoslo en práctica:

````python
import numpy as np

A = np.zeros((4,4))

print("A inicial =\n", A)
print()

A[2,:] = 3.1415
print("A luego de la primera asignación =\n", A)
print()

A[1::2, 1:3] = np.reshape(np.arange(1,5), (2,2))
print("A luego de la segunda asignación =\n", A)
````

<a name="indexacion_arreglos"></a>
### **2.4.2 Indexación usando listas o arreglos de enteros**
Otra opción posible a la hora de indexar arreglos es usar listas u otros arreglos. Por ejemplo, dada una matriz `A`, `A[ [1,3], [-4,-1] ]` selecciona los elementos en la 2da fila y trasantepenúltima columna junto con aquel en la 4ta fila y la última columna. Noten que esto selecciona los elementos `A[i,j]` con `(i,j)` donde `i` se toma de `[1,3]` y `j` de `[-4,-1]`. Naturalmente, para que la operación tenga sentido, ambas listas deben tener la misma cantidad de elementos. Resulta inmediato ver como esto generaliza a arreglos de mayor dimensionalidad.

Probemos esta forma de indexar con el ejemplo sugerido

````python
import numpy as np

A = np.reshape(np.arange(25, dtype=float), (5,5))

print("A =\n", A)
print()
print("A[ [1,3], [-4,-1] ] =\n", A[ [1,3], [-4,-1] ])
print()

A[ [1,3], [-4,-1] ] = np.random.random_sample((2))

print("A luego de la asignación =\n", A )
print()
````

Vemos que efectivamente, es una operación válida. También pudimos ver que nuevamente puede utilizarse esta forma de indexación para asignar sobre una porción del arreglo. Finalmente, noten que cumplimos lo que dijimos en [2.5.3](#opcionales) de consumir funciones que proveen argumentos opcionales. En particular, utilizamos el argumento `dtype` de `arange`. `dtype=float` especifica que deseamos que `arange` devuelva un arreglo de números de punto flotante (como ya hemos visto, por defecto, devuelve un arreglo de enteros).

Además de utilizar listas (o arreglos) para indexar un arreglo, podemos combinar dicha notación con el uso de rebanadas. De esta manera, `A[ [1,3,4], :]` nos permite acceder a la totalidad de los elementos de la 2ra, 4ta y 5ta fila de `A`. Adicionalmente, y probablemente ya no les sorprenda, también puede realizarse esta operación para asignar un subconjunto del arreglo.

````python
import numpy as np

A = np.reshape(np.arange(25, dtype=float), (5,5))

print("A =\n", A)
print()
print("A[ [1,3,4], : ]", A[ [1,3,4], :])
print()

A[ [1,3,4], :] = np.random.random_sample((3,5))
print("A después de la asignación =\n", A)
print()
````

<a name="indexacion_logica"></a>
### **2.4.3 Indexación con arreglos booleanos**

Finalmente la última opción que nos resultará interesante para seleccionar elementos o subconjuntos de arreglos es la indexacióón mediante listas o arreglos booleanos.

Para ello, una opción es utilizar una lista o vector de booleanos para cada dimensión a indexar, vector cuya longitud debe coincidir con la longitud del arreglo en esa dirección.

Por ejemplo, para seleccionar filas de una matriz $A$ de $m\times n$ elementos, precisaremos una lista o vector booleano con $m$ elementos. En particular, en los índices correspondientes a filas que deseamos retener, el vector debe tener un valor `True`, mientras que será `False` para aquellas que deseamos descartar. 

En este contexto, es usual denominar a los vectores de booleanos como **máscaras**, y al arreglo resultante, un arreglo enmascarado. Existe un submódulo de Numpy, llamado `np.ma`, dedicado a optimizar computacionalmente operaciones entre arreglos enmascarados. No veremos su uso en esta notebook.

Veamos un ejemplo de como indexar con listas/vectores de booleanos.

````python
import numpy as np

A = np.reshape(np.arange(25, dtype=float), (5,5))

filas    = [True, False, False, True, True]
columnas = np.array([True, False, True, True, False])

print("A =\n", A)
print()
print("Máscara para filas: ", filas)
print("A con sus filas enmascaradas :\n", A[filas, :])
print()
print("Máscara para columnas: ", columnas)
print("A con sus columnas enmascaradas:\n", A[:, columnas])
print()
print("A con sus filas y columnas enmascaradas:\n", A[filas, columnas])
````

Donde vimos, nuevamente, que al indexar una matriz (o, en general, un arreglo de dimensionalidad superior) utilizando vectores (o, en general, arreglos de dimensionalidad inferior), los elementos seleccionados son aquellos dados por las combinaciones de ambos vectores.

Otra posibilidad a la hora de indexar con booleanos, y que no debería sorprenderles luego del ejemplo anterior, es utilizar arreglos de dimensionalidad superior. Por ejemplo, es posible indexar una matriz utilizando una arreglo bidimensional de booleanos. Luego, la cantidad de elementos que serán retenidos al indexar serán aquellos en cuyos índices la máscara guarde el valor `True`. 

Resulta sencillo verlo a partir de un ejemplo

````python
import numpy as np

A         = np.reshape(np.arange(16, dtype=float), (4,4))
aleatorio = np.random.random_sample( (4,4) )
mascara   = np.zeros( (4,4), dtype=bool)

mascara[ [0, 0, 1, 3], [0, 1, 0, 2] ] = True

print("A =\n", A)
print()
print("Máscara para A:\n", mascara)
print()
print("A enmascarado:\n", A[mascara])
print("-----------------------")
print("Arreglo aleatorio:\n", aleatorio)
print()
A[mascara] = aleatorio[mascara]
print("A con los elementos enmascarados tomados del arreglo aleatorio:\n", A)
````

### **2.5.1 Operadores aritméticos**

Comencemos por considerar los operadores entre arreglos que generalizan aquellos vistos en [1.2](#aritmetica), es decir, la suma `+`, resta (`-`), multiplicación (`*`), división (`/`) y potenciación (`**`).

Para sumar el mismo número a un arreglo alcanza con expresar `v+num`, con `v` y `num` el arreglo y el número en cuestión, respectivamente. Se procede de manera análoga con el resto de los operadores.

Consideremos un ejemplo de esto. Sea $\vec v$ un vector con las velocidades instantáneas de un sistema de partículas en $\mathrm{km/s}$, calculemos $\vec v'$ la velocidad para un observador que se mueve a $U=0.5 \ \mathrm{km/s}$ respecto del primero y calculemos la energía cinética para cada partícula en Joules asumiendo que todas poseen una masa de $1 \ \mathrm{kg}$.

````python
import numpy as np

m  = 1                               # Masa de las particulas (kg)
v  = np.random.random_sample( (5) )  # Velocidades medidas (km/s)
U  = 0.5                             # Velocidad del otro observador (km/s)

vp      = v - U                      #  " medida por el otro observador (km/s)
vp_m_s  = vp*1000                    # Velocidad en m/s

E  = m*(vp_m_s**2)/2                 # Energía cinética de cada partícula (J)

print("Velocidades medidas (km/s):", v)
print("Velocidades que ve el observador primado (km/s):", vp)
print("Velocidades en el sistema primado (m/s):", vp_m_s)
print("Energía cinética de cada partícula (J):", E)
````

Como vemos, es bastante sencillo operar un arreglo y un número. 

Supongamos ahora que tenemos otro experimento donde medimos la velocidad de un sistema de partículas en un cierto instante utilizando dos técnicas diferentes, obteniendo sendos conjuntos de $\vec v_1$ y $\vec v_2$. Queremos ahora calcular la velocidad promedio $\bar{\vec v}$ de cada una de las partículas, dada por $\bar v^{(i)} = \big (v^{(i)}_1 + v^{(i)}_2 \big )/2$, con $v^{(i)}$ la velocidad de la i-ésima partícula.

Una manera fácil de hacerlo es utilizando los operadores que introdujimos anteriormente, entendiendo que cuando operan entre dos arreglos, lo hacen elemento a elemento (veremos al final de esta sección cómo el caso arreglo + número es un caso particular de esto). Es decir, siendo `u` y `v` arreglos, y tomando como ejemplo la suma, la instrucción
```python
w = u + v
```
es equivalente a 
```python
for i in range(N):
    w[i] = u[i] + v[i]
```
donde `N` es la cantidad de elementos de `u` (y `v`) y `w` es un vector de igual tamaño.

Utilicemos esto para calcular la velocidad promedio

````python
import numpy as np

v       = np.random.randint( 5, 8, (5) )  # Velocidad "real"

ruido1  = np.random.random_sample( (5) )  # Ruido en la medición 1
ruido2  = np.random.random_sample( (5) )  # Ruido en la medición 2

v1      = v + ruido1   # Medición 1
v2      = v + ruido2   # Medición 2

v_medio = (v1 + v2)/2  # v promedio

print("v1         = ", v1)
print("v2         = ", v2)
print("v promedio = ", v_medio)
````

Noten que introdujimos una nueva función, `randint(min, max, shape)` que devuelve un arreglo de números enteros al azar, tomados de una distribución uniforme entre `min` y `max` (no inclusivo) y con forma `shape`. De manera análoga podriamos usar cualquiera de las operaciones aritméticas mencionadas entre arreglos, operando elemento a elemento.

### **2.5.2 Operadores de comparación y lógicos**



<a name="numpy_comparacion"></a>
#### **2.5.2.1 Operadores de comparación**

Además de los operadores aritméticos, también podemos utilizar los comparadores vistos en [1.3](#logica), pero esta vez para operar sobre arreglos. En particular, los operadores `>`, `>=`, `<`, `<=`, `==` y `!=` (mayor, mayor o igual, menor, menor o igual, igual a o distinto a, respectivamente) pueden aplicarse elemento a elemento sobre arreglos, o entre un arreglo y un valor fijo.

Veamos un ejemplo utilizando vectores

````python
import numpy as np
#  Darle un valor fijo a random.seed permite tener reproductibilidad en los
# números pseudoaleatorios generados.
np.random.seed(10)

u = np.random.randint(5, 10, (5) )
v = np.random.randint(5, 10, (5) )

print("Vector de números al azar u:", u)
print("u >  7.5:", u > 7.5)
print("u == 6  :", u == 6 )
print()
print("----------------------------------")
print()
print("Vector de números al azar v:", v)
print("u == v:", u == v)
print("u != v:", u != v)
print("u >= v:", u >= v)
````

#### **<u>Comparando números de punto flotante</h5></u>**

Consideremos ahora una manera (un poco) más apropiada de comparar números de punto flotante (y números complejos). Como mencionamos en [1.3](#logica), llevarlo a cabo utilizando el operador `==` es una mala idea.

Una manera de hacerlo sencillamente en Python, y usando Numpy, es mediante la función `isclose(a, b, rtol=, atol=)`. Esta función determina si `a` y `b` son números de punto flotante o complejos cercanos entre sí usando la fórmula `|a-b| <= atol + rtol*|b|` por lo que `rtol` es una tolerancia relativa a la hora de comparar (escala con la magnitud de uno de los operandos) y `atol` una tolerancia absoluta. `atol` y `rtol` son argumentos opcionales y poseen valores sensatos cuando se comparan números que difieren pocos ordenes de magnitud de 1. Adicionalmente, `isclose` puede operar entre arreglos.

Veamos un ejemplo

````python
import numpy as np

a = np.array( [0.45+0.55, 0.1+0.2, 2.3-1.2] )
b = np.array( [1.0, 0.3, 1.1] )

print("Vector a:", a)
print("Vector b:", b)
print()
print("a == b         :", a==b)
print("np.isclose(a,b):", np.isclose(a,b))
````

<a name="numpy_logicos"></a>
#### **2.5.2.2 Operadores lógicos**

Si bien los operadores de comparación generalizan intuitivamente al trabajar con arreglos, no sucede lo mismo con los operadores lógicos `and`, `or`, `not`, como podemos ver en el siguiente ejemplo

````python
import numpy as np

a = np.array( [True, False, True] )
b = np.array( [False, False, True] )

print("Vector a:", a)
print("Vector b:", b)
print()
print("a and b         :", a and b)
````

Como pueden ver, la operación `and` no funciona para operar entre arreglos. Problemas similares se encuentran al utilizar `or` o `not`. Esto se debe a que la operación así formulada resulta ambigua, ¿queremos hacer una conjunción lógica elemento a elemento entre los arreglos, obteniendo un tercer vector booleano? ¿o queremos saber si hay algún valor verdadero en cada uno de los arreglos obteniendo un único valor booleano?

Si bien Numpy podría optar por darle alguno de los significados sugeridos al operador `and` (u otro diferente), los desarrolladores de la biblioteca han optado por hacer que `a and b` sea gramaticalmente incorrecto. En su lugar, se introducen operadores específicos que realizan las operaciones mencionadas previamente.

Para operar elemento a elemento, Numpy introduce 3 nuevos operadores operadores, la conjunción elemento a elemento (`&`), la disyunción elemento a elemento (`|`) y la negación elemento a elemento (`~`).

Veámoslos en funcionamiento

````python
import numpy as np

np.random.seed(1)

u = np.random.randint(5, 10, (5) )  > 7.5
v = np.random.randint(5, 10, (5) ) <= 8

print("Vector de booleanos u:", u)
print("Vector de booleanos v:", v)
print()
print("~ v  :",  ~ v)
print("u & v:", u & v)
print("u | v:", u | v)
````

Además de los operadores elemento a elemento, Numpy también provee las funciones `all` y `any` para reducir un arreglo booleano a un único valor de verdad. La primera de ellas devuelve `True` si y solo si todos los elementos del arreglo guardan el valor `True`, es decir, es la conjunción de todos los elementos del arreglo. La segunda, por su parte, realiza una disyunción a lo largo de todo el arreglo, devolviendo `True` con solo encontrar un elemento verdadero en el mismo.

Veamos un ejemplo de estos operadores en funcionamiento. Para ello consideremos un vector que contiene la cantidad de [títulos de _grand slam_](https://es.wikipedia.org/wiki/T%C3%ADtulo_de_Grand_Slam) que ha ganado une deportista, y donde cada índice denota un torneo distinto.

Analicemos, primero, si ha conseguido el _Grand Slam de carrera_ (i.e. ganar todos los torneos al menos una vez) y después si ha ganado algún torneo 4 o más veces.

````python
import numpy as np

np.random.seed(9)
titulos = np.random.randint(0, 5, (4) )

print("Cantidad de torneos de cada tipo ganados", titulos)
print("¿Consiguió un grand slam de carrera?", np.all(titulos > 0))
print("¿Ganó un torneo 4 o más veces?", np.any(titulos >= 4) )
````

<a name="broadcasting"></a>
### **2.5.3 _Broadcasting_**

Supongamos ahora que deseamos multiplicar cada fila de una matriz $A$ por un número distinto. Es decir, queremos multiplicar la $i$-ésima fila de $A$ por un número $u_i$. Intentemos hacer esta operación sin pensarlo demasiado.

````python
import numpy as np

A = np.random.randint(0, 4, (3, 2) )
u = np.random.randint(5, 10, (3) )

print(u*A)
````

Como vemos, Python no sabe como multiplicar $\vec u$ (el vector que contiene a los $u_i$) contra $A$. Cuando buscamos efectuar una operación que se define elemento a elemento (como `*` o `+`) entre dos arreglos que tienen distinta dimensionalidad, Numpy intenta seguir unas reglas, llamadas de **_broadcasting_** para realizar la operación.

El caso de _broadcasting_ más sencillo es el de multiplicar una matriz o un vector por un escalar. En ese caso la regla de _broadcasting_ indica que debe promoverse el escalar a la dimensión de la matriz (o del vector) y multiplicarse elemento a elemento ambas matrices, la original y la que tiene todos sus elementos iguales al escalar.

En general, al encontrarse con dos arreglos de diferente dimensionalidad, Numpy primero revisa que sean compatibles. Para verificar esto, se examinan de atrás para adelante las dimensiones de ambos arreglos, siendo compatibles dos dimensiones si, o bien coinciden, o una de ellas tiene longitud 1. De esta forma un arreglo de dimensiones $4\times 5\times 10$ es compatible con uno de la forma $5\times 1$, ya que $10$ es compatible con $1$ y $5$ con $5$. Asimismo, el arreglo de $4\times 5\times 10$ no resulta operable contra otro de dimensiones $4\times 10$, ya que $10$ es compatible con $10$ pero no $5$ con $4$. En caso que los arreglos a operar no sean compatibles Numpy reportará un error, como vieron más arriba.

Luego, si los arreglos son compatibles, lo que hace Numpy es _promover_ el arreglo más pequeño hasta que alcance el tamaño del de mayor dimensionalidad, como vimos que hacía con el escalar. De esta manera si tenemos un arreglo tridimensional $A$ (i.e. un _cubo_ de números) de dimensiones $4\times 3\times 3$ y lo multiplicamos por una matriz $B$ de $3\times 3$, lo que hará numpy es repetir $4$ veces a $B$ para llevarlo a la forma $4\times 3\times 3$ y luego operar elemento a elemento. Para entender un poco mejor como funcionan las reglas de _broadcasting_ pueden emplear las siguientes referencias en inglés: [_Broadcasting_](https://numpy.org/doc/stable/user/basics.broadcasting.html) y [_Array Broadcasting in Numpy_](https://numpy.org/doc/stable/user/theory.broadcasting.html#array-broadcasting-in-numpy).

Discutido como funciona el _broadcasting_, podemos usarlo ahora a nuestro favor para realizar la operación buscada: multiplicar cada fila de una matriz $A$ por un elemento de un cierto vector.

Para ello, sabemos ahora que alcanzará con cambiar la dimensionalidad de $\vec u$ para que sea $n \times 1$ en lugar de $n$. De esta manera, si $A$ era de $n\times m$, ahora $m$ será compatible con 1 y entonces Numpy promoverá $\vec u$ a una matriz de $n \times m$ donde cada columna contiene a $\vec u$. Luego, operará elemento a elemento.

Veamos que efectivamente es así.

````python
import numpy as np

A = np.random.randint(0, 4, (3, 2) )
u = np.random.randint(0, 4, (3) )

print("A =\n"  ,           A)
print("u ="    ,           u)
print("u*A =\n", u[:,None]*A)
````

Que efectivamente devuelve el resultado buscado. Noten que para agregar una nueva dimensión con longitud 1 a un arreglo utilizamos la palabra `None` en el lugar que correspondería indexar ese nuevo eje. Esto podría realizarse sobre más de un eje. Por ejemplo si $c$ es un vector de $n$ elementos y buscamos convertirlo en un arreglo tridimensional de dimensión $1 \times n \times 1$, bastará con notar `c[None,:,None]`.

Noten que de esta forma resulta súper sencillo definir el producto exterior entre dos vectores $A_{ij} = u_i v_j$ como

````python
import numpy as np

u = np.random.randint(1, 5, (3) )
v = np.random.randint(1, 5, (3) )

A = u[:, None] * v[None, :]

print("u ="  , u)
print("v ="  , v)
print("A =\n", A)
````

### **2.5.4 Operaciones entre vectores y matrices**

Naturalmente, las operaciones y funciones mencionadas anteriormente funcionan con arreglos de dimensionalidad arbitraria.

Sin embargo, esta subsección busca introducir operaciones cuyo uso más común es el de operar entre sí vectores, matrices o un vector y una matriz.

Estas operaciones incluyen al producto interno entre vectores, que podemos realizar usando la función `np.dot(vector1, vector2)` y a la multiplicación matricial, mediante `np.matmul(matriz1, matriz2)`.

Veamos como funcionan:



````python
import numpy as np

# Dot
vector1 = np.random.random_sample( (3) )
vector2 = np.random.random_sample( (3) )

producto_manual = vector1[0]*vector2[0] + vector1[1]*vector2[1] +\
                  vector1[2]*vector2[2]

print("Vector 1=", vector1)
print("Vector 2=", vector2)
print("Producto interno:", np.dot(vector1, vector2))
print("Lo compruebo:", producto_manual)
print()

# Matmul
matriz1 = np.random.random_sample( (2,2) )
matriz2 = np.random.random_sample( (2,2) )

elemento11 = matriz1[0,0]*matriz2[0,0] + matriz1[0,1]*matriz2[1,0]
elemento12 = matriz1[0,0]*matriz2[0,1] + matriz1[0,1]*matriz2[1,1]
elemento21 = matriz1[1,0]*matriz2[0,0] + matriz1[1,1]*matriz2[1,0]
elemento22 = matriz1[1,1]*matriz2[0,1] + matriz1[1,1]*matriz2[1,1]

print("Matriz 1=\n", matriz1)
print("Matriz 2=\n", matriz2)
print("Matriz 1 x Matriz 2=\n", np.matmul(matriz1, matriz2))
print("Lo compruebo:", elemento11, elemento12, elemento21, elemento22)
````

Como vemos, su uso es bastante sencillo. Adicionalmente, en versiones recientes de Python (mayores o iguales a la version 3.5), puede usarse el operador binario `@` como sinónimo de `np.matmul`. De esta manera, el producto entre dos matrices `A` y `B` puede realizarse como `A_por_B = A @ B`. En Google Colab van a poder utilizar esta notación sin problemas, sin embargo, ¡tengan cuidado al correr en computadores con versiones más antiguas de Python!

Si bien `np.dot` y `np.matmul` admiten argumentos de mayor dimensionalidad, las reglas que las generalizan no son del todo intuitivas y, sobre todo, no serán necesarias durante el curso. En caso que les interese construir operaciones arbitrarias del estilo multiplicación + reducción, pueden mirar la documentación de [`np.einsum`](https://numpy.org/doc/stable/reference/generated/numpy.einsum.html#numpy.einsum), función que tampoco usaremos durante el curso.

Otra operacion muy común al trabajar con matrices (que incluyen a los vectores fila y columna como casos particulares), es la operación de transposición. La misma puede realizarse mediante la función `np.transpose` o, más sencillamente, mediante el método `.T`, como muestra el siguiente ejemplo donde multiplicamos la transpuesta una matriz por un vector columna.

````python
import numpy as np

A = np.random.random_sample( (2,2) )  # Matriz
v = np.random.random_sample( (1,2) )  # Vector fila

print("A   = \n", A)
print("v   = \n", v)
print()
print("A.T = \n", A.T)
print("v.T = \n", v.T)

print("A.T x v.T", A.T @ v.T)

# Comentar la línea inferior después de ejecutarlo para suprimir el error
A.T @ v 
````

Vimos que es muy sencillo transponer arreglos. Además, como era de esperar, al querer hacer el producto matricial a derecha con una matriz fila, Numpy nos avisa que hay un error (comenten la última línea para ver que el error desaparece).

## **2.7 Operaciones sobre ejes (_`axis`_)**
([Regresar a la tabla de contenidos](#contenidos))

Como habrán visto en la enumeración anterior, una gran cantidad de funciones soportan el argumento opcional `axis` (o, por motivos bastante confusos, `axes` en `fft.fftn`). Este argumento indica que la operación debe realizarse sobre uno (en caso que `axis` sea un entero) o varios (en caso que `axis` sea una tupla de enteros) de los ejes del arreglo. 

Un eje es un término general para denotar fila, columna, etc. En un arreglo de $N\times M$ un índice que se mueve sobre el eje 0 corresponde a moverse a través de las filas, mientras que uno que lo hace sobre el eje 1 lo realiza sobre las columnas. Resulta inmediato entender como generaliza esto a arreglos de dimensión arbitraria.

De esta manera, dada una matriz `A` de `Nx3`, podemos obtener los `N` valores que surgen de calcular la norma de cada fila usando `np.mean(A, axis=1)` (noten que para obtener la norma de cada fila debemos movernos sobre las columnas al calcular cada resultado). Veamos que esto efectivamente es así.


````python
import numpy as np

A = np.random.random_sample( (4,3) )

print("Norma de cada fila, usando numpy:", np.linalg.norm(A, axis=1))

norm0 = np.sqrt(A[0,0]**2 + A[0,1]**2 + A[0,2]**2)
norm1 = np.sqrt(A[1,0]**2 + A[1,1]**2 + A[1,2]**2)
norm2 = np.sqrt(A[2,0]**2 + A[2,1]**2 + A[2,2]**2)
norm3 = np.sqrt(A[3,0]**2 + A[3,1]**2 + A[3,2]**2)

print("Norma de cada fila calculada manualmente:", norm0, norm1, norm2, norm3)
````

De manera análoga, si tuvieramos un arreglo tridimensional `T` (podría representar la temperatura en cada punto del espacio discretizado) con `NxMxL` elementos, y quisieramos calcular la temperatura media a cada altura, podríamos hacerlo con 
```python
promedio_horizontal = np.mean(T, axis=(0,1))
```
y de esta manera obtener nuestros `L` valores. Supusimos que `T` fue construido de tal manera que el eje 2 correspondía a movernos en altura.

Al comienzo puede resultar confuso razonar sobre cómo determinar qué valor de `axis` se busca. Una forma de pensarlo, es analizar en qué lugar habría que colocar el índice si recorrieramos el arreglo con un loop.

Considerando el ejemplo de las normas, podríamos calcular la norma de la 2da fila mediante un bucle de la siguiente manera

````python
norm = 0
for j in range(0, A.shape[1]):
    norm = norm + A[1, j]**2

print(np.sqrt(norm))
````

que como vemos coincide con lo hallado anteriormente. Noten que `j`, el índice sobre el que iteramos, ocupaba el lugar del eje 1 (recuerden que Python empieza a contar por 0).

## **2.8 Comparación de rendimiento**
([Regresar a la tabla de contenidos](#contenidos))

Una pregunta válida después del último ejemplo es por qué molestarse en usar `np.mean`, `np.linalg.norm`  o cualquier otra función similar, que implica recordar las reglas establecidas por los desarrolladores de Numpy (aunque resulten intuitivas), cuando son operaciones que resultan muy sencillas de programar uno mismo.

Una respuesta posible tiene que ver con evitar reinventar al rueda si no es necesario. Al programar nuestra propia función que realice alguna de estas tareas, podemos cometer errores, mientras que las funciones que provee Numpy han sido usadas extensamente y, por tanto, es menos probable que presenten errores. No obstante, cuando sea por una razón específica, implementar una versión propia puede ser una opción a considerar.

Otro motivo bastante importante para usar las funciones provistas por Numpy es que las mismas suelen tener un nivel de optimización que no puede alcanzarse programando solo en Python. Estas funciones se dice que están _vectorizadas_. Sin entrar en detalles de implementación, esto quiere decir que aprovechan más eficientemente la distribución de los arreglos en la memoria RAM de la computadora (y en los registros de la CPU). Además, bajo condiciones apropiadas, las operaciones vectorizadas pueden realizar varias operaciones elemento a elemento en la misma instrucción de procesador (CPU), usando lo que se denominan instrucciones [SIMD](https://es.wikipedia.org/wiki/SIMD).

Veamos la diferencia de rendimiento entre una implementación manual y otra de Numpy. Para ello ejecutan las siguientes 3 celdas.

````python
import numpy as np
N = 10000
A = np.random.random_sample( (N,3) )
````

````python
%%time
resultado_numpy  = np.linalg.norm(A, axis=1)
````

````python
%%time
resultado_manual = np.zeros( (N) )
for i in range(N):
    for j in range(3):
        resultado_manual[i] = resultado_manual[i] + A[i,j]**2
    resultado_manual[i] = np.sqrt(resultado_manual[i])
````

````python
# Comprobemos que dieron lo mismo
print("¿Los valores son equivalentes?",
      np.all(np.isclose(resultado_numpy, resultado_manual)))
````

Si bien esta no es la manera óptima de medir el rendimiento de cada método, la diferencia entre las implementaciones es tan significativa que no será muy sensible a cómo la midamos. Cuando corran estas celdas verán un factor de entre 50 y 100 en el tiempo de ejecución total (_Wall time_) de cada método.

## **3.1 Figuras y pares de ejes**
([Regresar a la tabla de contenidos](#contenidos))


Luego de, naturalmente, _importar_ `matplotlib.pyplot`, lo primero que vamos a necesitar para realizar los gráficos que nos interesan es generar un lienzo y dividirlo en tantas secciones como cosas cualitativamente distintas busquemos visualizar. Hay varias formas de realizar esta operación, sin embargo la que considero más sencilla y fácil de generalizar a casos más complejos es usando la función `matplotlib.pyplot.subplots`.

Veamos como funciona

````python
import matplotlib.pyplot as plt

fig, axs = plt.subplots(3, 2, figsize=(9,4), constrained_layout=True)
````

Como vemos, esto  genera una figura (como denominaremos a partir de ahora al lienzo) que contiene a 6 pares de ejes $(x,y)$. En particular, invocamos a `subplots` con dos argumentos obligatorios: el primero (`3` en nuestro caso) es la cantidad de pares de ejes apilados verticalmente, mientras que el segundo (`2` en nuestro caso) es la cantidad apilada horizontalmente. Además utilizamos dos argumentos opcionales (de los muchos que hay). `figzise=(H,V)` que recibe una tupla de enteros cuyo primer elemento `H` es la extensión horizontal de la figura mientras que `V` es longitud vertical de la misma. Por otra parte, y dicho livianamente, `constrained_layout=True` hace que Matplotlib haga un esfuerzo extra para intentar que los elementos de un par de ejes no se superpongan con los de otro.

Como habrán notado, las divisiones (i.e. los pares de ejes) obtenidos son todas de igual tamaño. Pueden lograrse disposiciones más complejas usando [`gridspec`](https://matplotlib.org/stable/tutorials/intermediate/gridspec.html), aunque no lo veremos aquí.

Como vemos, `subplots` de vuelve dos cantidades, que llamamos `fig` y `axs`. `fig` es una referencia a la figura creada y entonces tareas que queramos realizar para toda la figura, como por ejemplo poner un título global, se accederán a través de `fig`. Por su parte `axs` es un referencia a todos los pares de ejes creados. En particular, `axs` es un arreglo Numpy cuya dimensionalidad depende de cuantos pares de ejes creamos. En nuestro caso, que tenemos pares de ejes apilados tanto horizontal como verticalmente, `axs` será un arreglo bidimensional de forma `(3,2)` y, por ejemplo, `axs[2,0]` referirá al par de ejes ubicado en la última fila y la primer columna (abajo a la izquierda).

## **3.2 Gráficos simples (`plot`)**
([Regresar a la tabla de contenidos](#contenidos))


Teniendo ya un espacio sobre el que trabajar, estamos ahora en condiciones de crear visualizaciones. Para comenzar generemos un gráfico de alguna función sencilla, por ejemplo, de un seno y un coseno. Esto quiere decir que vamos a graficar pares $\left[x, \mathrm{sen}\left(x\right)\right]$ y $\left[x, \cos\left(x\right)\right]$. Para esto usaremos el método `plot` sobre un par de ejes.

````python
import numpy as np
import matplotlib.pyplot as plt

x   = np.linspace(0, 2*np.pi, 1000)  # 1000 puntos en [0, 2pi]
sin = np.sin(x)
cos = np.cos(x)

# Creo la figura y el par de ejes
fig, ax = plt.subplots(1,1, figsize=(8,3))
ax.plot(x, sin)
ax.plot(x, cos);
````

Como vemos, resulta muy sencillo de utilizar. Alcanza con pasarle a `plot` la coordenada $x$ de cada punto que queremos graficar como primer argumento y la coordenada $y$ de los mismos como segundo argumento. Además, para realizar varios gráficos en el mismo par de ejes alcanza con llamar sucesivamente a `plot` (o a cualquier otro método para gráficar) de manera sucesiva y Matplotlib los agrega en el orden en que fueron generados. Asimismo, cuando hay más de un elemento a graficar, Matplotlib va asignándole un color distinto a cada uno (azul para el seno y naranja para el coseno, en nuestro caso).

### **3.2.1 Escala logarítmica (`loglog` y `semilog?`)**

Como seguramente sepan a esta altura de sus carreras, una escala lineal en los ejes no siempre resulta apropiada para lograr una buena visualización. Hay relaciones que se ven más claramente graficando, por ejemplo, pares $\left[\log(x), \log\left(f(x)\right)\right]$. Sin embargo, al graficar $\log\left(f(x)\right)$ en escala lineal puede ser difícil apreciar visualmente cambios cuantitativos en $f(x)$. Para esto podemos emplear gráficos en escala logarítmica.

Un caso donde una escala logarítmica es particulamente útil es aquel donde la relación que vemos sigue una ley de potencias. Por ejemplo, la tercera ley de Kepler establece que $T \propto r^{3/2}$ para todos los cuerpos que describen una órbita elíptica alrededor del mismo objeto primaro, y donde $T$ es el período de la órbita y $r$ el semi-eje mayor de la órbita. Veamos como se ve esta relación en escala lineal y como verla en escala logarítmica.

````python
import numpy as np
import matplotlib.pyplot as plt

r = np.linspace(1, 50, 1000)
T = r**1.5

fig, axs = plt.subplots(1, 2, figsize=(8,3), constrained_layout=True)
axs[0].plot(r, T)     # Escala lineal
axs[1].loglog(r, T);  # Escala logarítmica 
````

Como vemos, mientras que en escala lineal la relación tiene una cierta curvatura difícil de estimar, en escala logarítmica la relación se ve sencillamente como una recta con pendiente $3/2$. Más aún, vimos que resulta súper sencillo graficar en escala logarítmica, basta con reemplazar `plot` por `loglog`. Adicionalmente, si se desea que solo uno de los ejes tenga escala logarítmica puede usarse `semilogx` para que la misma aplique solo sobre el eje $x$ (i.e. horizontal) o `semilogy` para que haga lo propio solo sobre $y$ (i.e. el eje vertical).

Vale resaltar que esta no es la única manera de establecer una escala logarítmica en alguno (o ambos) ejes, sin embargo creo que es la más intuitiva y compacta.

### **3.2.2 Puntos en lugar de curvas suaves (`marker` y `linestyle`)**

Como vimos más arriba, Matplotlib recibe conjuntos discretos de puntos y genera curvas en apariencia continuas y suaves. Esto es razonable cuando se tiene una gran densidad de puntos (arriba usamos 1000 puntos) que varían suavemente. Sin embargo, cuando se tienen unos pocos datos, o los mismos contienen ruido, es una mejor idea graficar estrictamente los puntos medidos y no conectarlos de ninguna manera.

Para esto podemos usar los argumentos opcionales `marker` y `linestyle`, tanto en `plot` como en `loglog`, `semilogx` o `semilogy`. `marker` recibe un string que denota como marcar los puntos. Una lista completa de los marcadores soportados se encuentra en este [enlace](https://matplotlib.org/stable/api/markers_api.html). Asimismo `linestyle` denota como graficar la línea, y sus opciones pueden verse [aquí](https://matplotlib.org/stable/gallery/lines_bars_and_markers/linestyles.html). Notablemente, en dicha documentación no menciona explícitamente que puede usarse `""` para denotar que no se desea ningún tipo de línea. Veamos un ejemplo de esto

````python
import numpy as np
import matplotlib.pyplot as plt

# Distancias de los planetas del sistema solar
r     = np.array([0.39, 0.723, 1, 1.524, 5.203, 9.539, 19.18, 30.06])
T     = r**1.5                             # Relación de Kepler
ruido = np.random.normal(0, 0.5, r.shape)  # Ruido blanco

T_con_ruido = T + ruido*r**(3/2)  # "Mediciones"

fig, ax = plt.subplots(1, 1, figsize=(5,3))
ax.loglog(r, T_con_ruido, marker="x", linestyle="")
ax.loglog(r, r**(3/2), linestyle="--");
````

Esto podemos hacerlo con el método `text`, que recibe tres argumentos. Los primeros dos corresponden a la coordenada $x$ (primero) e $y$ (segundo) donde realizaremos la anotación y el tercero será un string que contenga al texto  con el que queremos anotar el gráfico.

Vale mencionar que para todos los elementos de texto Matplotlib soporta el uso de $\mathrm{\LaTeX}$. Para ello basta con encerrar el texto (que recuerden, debe ser un string) entre signos `$` (i.e. `$código LaTeX$`).

## **3.5 Mapas de calor (`imshow`)**
([Regresar a la tabla de contenidos](#contenidos))




Otro tipo de visualización frecuente en ciencias son los mapas de calor. Supongamos que tenemos un campo escalar bidimensional, por ejemplo la temperatura de una placa discretizada sobre `NX x NY` puntos de grilla, podemos representarla mediante un arreglo `T` de dimensiones `(NX, NY)`. Luego para ver visualmente la temperatura en toda la placa al mismo tiempo, podemos graficar para cada punto del espacio un punto cuyo color depende de la temperatura allí.

Para este propósito podemos usar el método `imshow`. `imshow` grafica una matriz, colocando por defecto al elemento `[0,0]` en la esquina superior derecha y a los elementos `[-1,0]` en la esquina inferior izquierda y al `[-1,-1]` en la inferior derecha. Veamos como funciona `imshow`:

````python
import numpy as np
import matplotlib.pyplot as plt

A = np.array( [ [1,2], [3,4] ] )
print("A=\n", A)
fig, ax = plt.subplots(1,1)
ax.imshow(A, cmap="binary");
````

Como ven, `imshow` soporta un argumento opcional `cmap` que permite seleccionar un mapa de colores. En este caso elegimos `binary` que equivale a colorear usando una escala de grises, con blanco para el valor más pequeño y negro para el mayor (de esta manera nos ahorramos, por ahora, usar una barra de colores).

Sin embargo, generalmente ocurre que dado un campo `T`, `T[0,0]` representa el valor que se halla abajo (no arriba como en nuestro ejemplo) y a la izquierda. Es por ello que `imshow` tiene la opción `origin`, para especificar si el origen debe situarse arriba (`origin=upper`) o abajo (`origin="lower"`).

Adicionalmente noten que, en general, `T[:,0]` refiere a puntos con coordenada `y` fija y que, por tanto, se desarrollan en dirección horizontal. Por otro lado, en `imshow` esto representa una columna, por lo que si queremos mantener al eje `x` como horizontal e `y` como vertical, debemos graficar la transpuesta de `T`. Veamos como sería la temperatura de una placa a temperatura con temperatura $T_0$ en $(x,y)=(0,0)$ y temperatura $T_1$ en $(x,y)=(\pi,1)$. Agreguémosle también un poco de ruido.

````python
import numpy as np
import matplotlib.pyplot as plt

T0, T1 = 1, 2

x = np.linspace(0, np.pi, 100)
y = np.linspace(0, 1, 100)

T = (T1-T0)*x[:,None]*y[None,:]/np.pi + T0
T = T + np.random.normal(0, 0.05, T.shape)

fig, ax = plt.subplots(1,1)
ax.imshow(T.T, origin="lower", extent=(x[0], x[-1], y[0], y[-1]));
````

Donde, como verán, utilizamos el argumento opcional `extent=(xmin, xmax, ymin, ymax)` para indicar las dimensiones del dominio.

### **3.5.1 Barras de color (`colorbar`)**

Un problema con el gráfico que generamos más arriba es que no podemos saber fácilmente a qué corresponden los colores amarillo y violeta (ni ninguna de las tonalidades en el medio). Para eso podemos usar una barra de colores, que es muy sencilla de generar usando `colorbar`.

Una manera (bastante general) de generar una barra de colores es obteniendo primero una referencia al gráfico cuyos colores queremos denotar, para esto podemos sencillamente asignar el gráfico a una variable usando
```python
graf = plt.imshow(...)
```
y luego usando `fig.colorbar` de la siguiente manera

````python
import numpy as np
import matplotlib.pyplot as plt

T0, T1 = 1, 2

x = np.linspace(0, np.pi, 100)
y = np.linspace(0, 1, 100)

T = (T1-T0)*x[:,None]*y[None,:]/np.pi + T0
T = T + np.random.normal(0, 0.05, T.shape)

fig, axs = plt.subplots(1,2, figsize=(12,3), constrained_layout=True)
axs[0].plot(T[x.size//2], y)
graf = axs[1].imshow(T.T, origin="lower", extent=(x[0], x[-1], y[0], y[-1]))

fig.colorbar(graf, ax=axs[1], orientation="horizontal");
````

Noten que graficamos nuestro mapa de color a la derecha, mientras que a la izquierda graficamos un corte vertical de la temperatura en la mitad del dominio. Pueden ver que, por un lado, `colorbar` es un método de la figura `fig` y no de algún par de ejes en particular, a diferencia de todos los que vimos hasta ahora. Es por ello que para indicar a qué par de ejes corresponde la barra de colores que creamos le pasamos el argumento opcional `ax` como `ax=axs[1]` (ya que queremos hacer una barra de colores en el segundo par de ejes). Además le pasamos la opción `"horizontal"` al argumento opcional `orientation` ya que por defecto las barras se colores de dibujan verticalmente.

## **3.6 Estilizar ejes y figuras**
([Regresar a la tabla de contenidos](#contenidos))

Aún cuando Matplotlib tiene muchísima más funcionalidad que la expuesta más arriba, el último tema que vamos a tratar es el de cómo estilizar (algunos) de los elementos de nuestros gráficos. En particular, vamos a ver cómo etiquetar los ejes, poner un límite a las secciones del gráfico a mostrar, añadir un grillado, generar leyendas y poner títulos.

**<h3>Etiquetar ejes</h3>**

Para etiquetar ejes pueden usar los métodos `ax.set_xlabel(str)` o `ax.set_ylabel(str)` para etiquetar los ejes $x$ (horizontal) e $y$ (vertical) respectivamente, y donde `ax` es el par de ejes con el que están trabajando. `str` es el texto que queremos colocar en la etiqueta.

**<h3>Colocar títulos</h3>**

Para colocar un título a un par de ejes pueden usar el método `ax.set_title(str)`.

**<h3>Delimitar una región del gráfico</h3>**

Para delimitar una región del gráfico pueden usar los métodos `ax.set_xlim(left=xmin, right=xmax)` y `ax.set_ylim(bottom=ymin, top=ymax)`, para establecer el límite izquierdo (`xmin` en este caso), derecho (`xmax`) del eje $x$ y los límites inferior (`ymin`) y superior (`ymax`) del eje $y$. En caso que no especifiquen alguno de los argumentos, Matplotlib intentará elegir inteligentemente un límite apropiado de forma de que el gráfico no se vea recortado en esa dirección (como vimos hasta ahora, donde no usamos ninguno de estos métodos).

**<h3>Colocar una grilla</h3>**

Para facilitar la legibilidad de valores lejos de los ejes, podemos agregar una grilla. Esto puede hacerse sencillamente con el método `ax.grid()` sobre el par de ejes deseado.

**<h3>Agregar una leyenda</h3>**

Para agregar una leyenda el procedimiento en cuestión consta de dos partes. Por un lado, al graficar una cierta cantidad (por ejemplo, usando `ax.plot`), hay que especificar el texto correspondiente a la leyenda usando el argumento opcional `label` (_etiqueta_ en inglés). Por ejemplo, `label="$\cos(x)$"`.

Luego, una vez que todos los objetos que se desean incorporar a la leyenda ya tienen su `label` asignado, basta con utilizar el método `ax.legend()`. A su vez, legend soporta los argumentos opcionales `ncol` (que especifica el número de columnas) y `loc` que especifica la ubicación de la leyenda. Pueden consultar los valores posibles para `loc` como así también otros argumentos que especifican el tamaño de letra, el color, etc. [acá](https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.legend.html).

**<h3>Usando todo junto</h3>**

Busquemos ahora usar lo que acabamos de ver para generar una versión más completa del gráfico que realizamos al final de la sección [**3.5**](#colorbar).

````python
import numpy as np
import matplotlib.pyplot as plt

T0, T1 = 1, 2

x = np.linspace(0, np.pi, 100)
y = np.linspace(0, 1, 100)

T = (T1-T0)*x[:,None]*y[None,:]/np.pi + T0
T = T + np.random.normal(0, 0.05, T.shape)

fig, axs = plt.subplots(1,2, figsize=(12,3), constrained_layout=True)
axs[0].plot(T[x.size//2], y, label="$T$ en la mitad de la placa")
axs[0].legend()
axs[0].grid()
axs[0].set_ylim(bottom=0.5)
axs[0].set_title("Temperatura en $x=\pi/2$ para y>0.5")
axs[0].set_xlabel("$T$")
axs[0].set_ylabel("$y$")

graf = axs[1].imshow(T.T, origin="lower", extent=(x[0], x[-1], y[0], y[-1]))
axs[1].set_xlabel("$x$")
axs[1].set_ylabel("$y$")

cbar = fig.colorbar(graf, ax=axs[1], orientation="horizontal")
cbar.set_label("$T$")
````

Donde adicionalmente vimos también que `colorbar` devuelve una referencia a la barra de colores creada (que asignamos a `cbar`) y que la misma admite configurar su etiqueta (en este caso $T$) mediante el método `cbar.set_label(str)`.

## **4.2 Guardar y cargar datos con Numpy**
([Regresar a la tabla de contenidos](#contenidos))

Veremos a continuación dos grandes maneras provistas por Numpy para guardar y leer datos. La primera de ellas es guardar datos como un archivo de texto, legible en cualquier editor de texto (como Vim o Notepad). La segunda, bastante más eficiente en cuanto a uso de almacenamiento, implica guardar los datos de manera binaria (y por tanto más difícil de acceder).

En ambos casos, el archivo que vamos a guardar va a tener que tener un nombre. De forma general, un nombre va a ser un string de la forma
```python
fname = "/dir1/dir2/dir3/dir4/nombre_de_archivo.ext"
```
donde `dir1`, `dir2`, etc. son los distintos directorios (o carpetas, si están acostumbrados a la nomenclatura de Windows). En una computadora local podría tener, por ejemplo,
```python
fname = "/home/mfontana/metodos_numericos/mi_archivo.ext"
```
mientras que en Google Colab, si quiero guardar el archivo en la carpeta `fundamentos_de_python_cientifico` dentro de `metodos_numericos` voy a tener que construir el string de mi archivo como
```python
fname = "/content/gdrive/My Drive/metodos_numericos/fundamentos_de_python_cientifico/mi_archivo.ext"
```

En la discusión anterior, `ext` es la extensión del archivo y, suele asociarse al tipo de archivo que guardemos o carguemos. En el caso de archivos de texto será `txt`, mientras que para los achivos binarios puede ser `npy`, `npz` o cualquier otra.

<a name="numpytxt"></a>
## **4.2.1 Archivos de texto**
Para guardar un arreglo como un archivo de texto (también llamado, a veces, archivo ASCII), podemos usar la función `np.savetxt(fname, arreglo)`, mientras que `np.loadtxt(fname)` nos servirá para cargar el mismo. Adicionalmente, `savetxt` posee el argumento opcional `header` (palabra del idioma inglés para _encabezado_) que nos permite poner un string en la primera línea y luego escribir el arreglo. Los encabezados son útiles para recordar que se guardó en cada fila.

Supongamos que el arreglo `datos` contiene un arreglo de 3 columnas, donde la primera es la cantidad de puntos en una simulación, el segundo el error de la misma y el tercero el tiempo de ejecución en segundos. Podemos escribir `datos` a un archivo de texto como
```python
fname = "/content/gdrive/My Drive/metodos_numericos/fundamentos_de_python_cientifico/resumen_simulaciones.txt"
encabezado = "Puntos | Error | Tiempo de ejecución (s)"
np.savetxt(fname, datos, header=encabezado)
```

Lo que generará un archivo con los siguientes contenidos
```
# Puntos | Error | Tiempo de ejecución (s)
128 1.7135781e-5 5
256 4.47178314-8 20
512 3.12873137-8 80
...
```

Luego este archivo podemos cargarlo como
```python
fname = "/content/gdrive/My Drive/metodos_numericos/fundamentos_de_python_cientifico/resumen_simulaciones.txt"
datos = np.loadtxt(fname)
```

En ambos casos, nuestros datos aparecerán separados por espacios en blanco `␣`. Si queremos usar otro carácter para separar los datos, podemos usar tanto en `savetxt` como en `loadtxt` el argumento opcional `delimiter` (`delimiter=","`, por ejemplo, para separar valores con `,`).

<a name="numpybin"></a>
## **4.2.2 Archivos binarios**

Para guardar un único arreglo como un archivo binario, Numpy provee las funciones `np.save(fname, arreglo)` y `np.load(fname, arreglo)`. Estas funciones guardan con formato `npy` (por lo que tiene sentido usar `.npy` como extensión).

**<h4>Un único arreglo (npy):</h4>**

Por ejemplo, si `T` es un arreglo que contiene un campo de temperaturas, podemos guardarlo como `temperatura.npy` de la siguiente manera
```python
fname = "/content/gdrive/My Drive/metodos_numericos/fundamentos_de_python_cientifico/temperatura.npy"
np.save(fname, T)
```
y luego cargarlo como
```python
fname = "/content/gdrive/My Drive/metodos_numericos/fundamentos_de_python_cientifico/temperatura.npy"
T = np.load(fname)
```

**<h4>Múltiples arreglos (npz):</h4>**

En caso que queramos guardar más de un arreglo en el mismo archivo, Numpy posee el formato `npz`, que guarda un diccionario de arreglos. Para ello guardar archivos `npz` se utiliza la función `np.savez(fname, arr1=arreglo1, arr2=arreglo2, ...)`, donde `arr1` es el nombre que queremos darle a los datos presentes en `arreglo1` y lo mismo para `arr2...arrn`. Por otra parte, para cargarlos se usa  `np.load(fname)` al igual que antes.

Para pensar en un ejemplo concreto, supongamos que `vx` y `vy` contienen las componentes del campo de velocidades de un fluido. Podemos guardar estos arreglos en un único archivo como
```python
fname = "/content/gdrive/My Drive/metodos_numericos/fundamentos_de_python_cientifico/velocidad.npz"
np.savez(fname, velocidad_x=vx, velocidad_y=vy)
```
y cargarlos luego utilizando
```python
fname = "/content/gdrive/My Drive/metodos_numericos/fundamentos_de_python_cientifico/velocidad.npz"
datos = np.load(fname)
vx = datos["velocidad_x"]
vy = datos["velocidad_y"]
```

# **Ejercicio integrador: inestabilidad de Kevin-Helmholtz**
([Regresar a la tabla de contenidos](#contenidos))

Como ejercicio final les proponemos ahora que intenten reproducir el gráfico usado como ejemplo en el comienzo de la [Sección 3](#matplotlib). Los distintos paneles que deben generar tienen distintos grados de dificultad, e inclusive hay elementos estéticos (como el uso de Computer Modern para renderizar LaTeX o la selección de un espaciado particular entre marcadores) que no fueron presentadas aquí, para aquelles que quieran revisar la documentación de Matplotlib o buscar respuestas en [Stack Overflow](https://stackoverflow.com).

La idea es estudiar un tipo de inestabilidad muy comun en fluidos, llamado inestabilidad de Kelvin-Helmholtz, que se produce cuando hay un fuerte gradiente de transversal en alguna componente del campo de velocidades. Es por ello que, partiendo de la siguiente situación inicial

<center></img></center>

donde vemos un campo de velocidades cuyo signo cambia abruptamente, lo evolucionamos en el tiempo luego de introducir una perturbación de magnitud despreciable. El dominio es bidimensional, siendo periódico en la dirección $x$ y tiene dos tapas rígidas estacionarias en $y=0$ e $y=2$. Luego de un cierto tiempo, se obtienen las componentes del campo de velocidades $\mathbf{u}$ y de la vorticidad $\mathbf{\omega} = \mathbf{\nabla} \times \mathbf{u}$ que se encuentran en el archivo `npz` disponible en [este](https://nube.df.uba.ar/index.php/s/srw3P2SBZeWmNEr) enlace. Usando estos datos, deben regenerar el gráfico mencionado, repetido aquí por practicidad.

<center></img></center>


El primer panel muestra la única componente no nula del campo de vorticidad $\omega_z$ para todo el dominio, con una barra de colores que permite estimar el valor de la vorticidad en una cierta región del espacio. Además se resaltaron los puntos ubicados sobre $y=1$ (es decir, el centro del canal) con una línea roja punteada, ya que allí se centrarán los estudios de la fila inferior. Por otro lado, el segundo panel (en la fila de arriba), muestra los perfiles medios verticales para las componentes no nulas de cada campo, una con cada color. Noten que
\begin{equation}
\langle a \rangle_x (y)= \frac{1}{N_x} \sum_{i=j}^{N_x} a(x_j, y),
\end{equation}
es el promedio simple de $a$ a lo largo de la dirección $x$.

El panel inferior izquierdo, muestra el espectro de potencias de la transformada de Fourier en la dirección $x$ para cada campo es decir
\begin{equation}
    |\hat{a}_m (y)|^2 = \left| \sum_{j=1}^{N_x} a(x_j, y) e^{- i 2\pi j m /N_x} \right|^2,
\end{equation}
con $[k_x]_m = \pi(2m-N_x)/L_x$, con $N_x$ la cantida de puntos y $L_x$ la longitud del dominio, ambas cantidades en la dirección $x$ (se consideró $N_x$ par).
En particular, se muestran cortes para $y=1$, es decir para el centro de la caja. Noten que para realizar este cálculo pueden serles de utilidad las funciones `np.fft.fft(axis=)` (para construir las ordenadas) y `np.fft.fftfreq` (para construir las abscisas). Adicionalmente, una teoría fenomenológica para la turbulencia, formulada por [Kolmogorov en 1941]((https://en.wikipedia.org/wiki/Turbulence#Kolmogorov's_theory_of_1941) (sepan disculpar la referencia en inglés), propone que en caso que la turbulencia sea isótropa y homogénea debe observarse un rango $k^{-5/3}$ en el espectro de potencia del campo de velocidades. Marcamos con una línea roja punteada y una etiqueta textual, un rango presuntamente compatible con dicha hipótesis.

Otro elemento que puede utilizarse para caracterizar la turbulencia, son las denominadas funciones de estructura, que es lo que graficamos en el último panel. Estas cantidades se construyen a partir de las componentes tubulentas de cada campo, que denotaremos con $'$ y que dada la anisotropía del problema podemos definir como
\begin{equation}
a'(x,y) = a(x,y) - \langle a \rangle_x (y).
\end{equation}
Luego, las funciones de estructura normalizadas de segundo orden del campo $\mathbf{a}$ se definen como
\begin{equation}
S^{2 \mathbf{a}}_{j\ell} (r_\ell, y) = \frac{ \left \langle \left[ a_j'(\mathbf{x} + r_\ell  \hat{\mathbf{n}}_\ell) - a_j'(\mathbf{x}) \right]^2 \right \rangle_x}{\langle |a_j|^2 \rangle_x},
\end{equation}
donde $\mathbf{x}$ y $\hat{\mathbf{n}}_\ell$ representan el vector posición y un versor en la dirección $\hat{\mathbf{x}}$ ($\ell=1$), $\hat{\mathbf{y}}$ ($\ell=2$) o $\hat{\mathbf{z}}$ ($\ell=3$). $S^2$ es entonces una medida de la diferencia entre el campo original y el campo desplazado en la dirección dada por $\hat{\mathbf{n}}_\ell$ y $r_\ell$ es la magnitud de dicho desplazamiento. Puede serles de utilidad `np.roll(axis=)`.
